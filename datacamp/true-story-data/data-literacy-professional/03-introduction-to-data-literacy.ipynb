### CHAPTER 1 — Data Literacy Basics

```
DATA LITERACY
│
├── Definition
│   └── The ability to READ, WORK WITH, ANALYZE, and COMMUNICATE insights with data
│       ├── Reading data        → identify sources, collect, manage
│       ├── Working with data   → turn data into insights (analytics)
│       └── Communicating       → visualizations, storytelling
│
├── DIKW Pyramid (Data → Information → Knowledge → Wisdom)
│   ├── Data        → raw observations, unorganized, no meaning yet
│   │                 (e.g. thermometer shows 15°C, dark clouds)
│   ├── Information → data placed into context, organized/aggregated
│   │                 (e.g. temp dropped → clouds formed → rain started)
│   ├── Knowledge   → combine information, detect patterns, generalizations
│   │                 (e.g. temp drop + dark clouds predicts rain)
│   └── Wisdom      → applied knowledge to act proactively
│                     (e.g. I know when rain comes → I bring umbrella)
│
│   > Insights are the key output between Knowledge and Wisdom
│
└── Data-Driven Decision Making
    ├── Definition: using data to make informed decisions and act on them
    ├── Benefits: optimize performance · better understanding · protect against risks
    ├── Misconceptions:
    │   ├── NOT only for large organizations
    │   ├── NOT only the data team's responsibility
    │   └── NOT always the best or only answer
    ├── 5-step data-driven process:
    │   Define problem → Collect data → Analyze data → Draw insights → Take action
    └── Problem statement qualities: clearly defined · actionable · realistic
        Categories: describe state · diagnose causes · detect anomalies / predict
```

### CHAPTER 2 — Reading Data

```
READING DATA
│
├── Data Sources
│   ├── Open data   → public databases (data.gov, data.europe.eu), public APIs
│   │                 (Twitter, Google Maps, IMDB, Wikipedia)
│   └── Internal data → customer data, transaction data, website data, survey data
│
├── Data Types
│   ├── Structured vs. Unstructured
│   │   ├── Structured    → tabular, easy to search, relational databases
│   │   │                   (spreadsheets, data tables)
│   │   └── Unstructured  → no pre-defined structure, harder to search
│   │                       (images, video, sound, text)
│   │
│   └── Quantitative vs. Qualitative
│       ├── Quantitative  → numbers, measurable, wider analysis range
│       │                   (count of scoops, minutes, length)
│       └── Qualitative   → categories, observable, restricted analysis range
│                           (flavor, color, cone or cup)
│
├── Managing Data
│   ├── DBMS: software for managing databases (data engineer responsibility)
│   ├── Database types: relational vs. document · data warehouse vs. data lake
│   ├── Cloud storage: remote servers, cost-effective, third-party security dependency
│   ├── Pipelines: ETL (Extract → Transform → Load) — automated, keeps data up-to-date
│   ├── Querying: SQL to access/retrieve data; R or Python for further analysis
│   └── Dashboards: KPIs at a glance, linked to database, created by data analyst
│
└── Common Data Problems (Dirty Data)
    ├── Dirty data = incorrect · incomplete · inconsistent
    │   Cause: human error · technical issues · collection process
    │   Principle: Garbage In, Garbage Out → wrong conclusions
    │
    ├── Data errors    → incorrect/inconsistent values or formats
    │   Counter: correct if original known; drop if unknown
    ├── Missing data   → incomplete records
    │   Problematic when: many missing · underlying patterns in missing data
    │   Counter: drop data OR imputation
    └── Data bias      → societal bias reflected in data → unrepresentative results
        Hard to detect and resolve
        Counter: sound collection process · awareness in conclusions · explainable AI
```

> **Cleaning rule:** Not all data problems are fully solvable — but it is always possible to do *some* kind of analysis.

### CHAPTER 3 — Working with and Analyzing Data

```
4 TYPES OF ANALYTICS
│
├── Descriptive   → WHAT happened?
│   ├── Goal: get to know data, investigate relationships, prep for advanced techniques
│   ├── Techniques: descriptive statistics · visualizations · outlier detection · EDA
│   └── EDA: assess characteristics · find patterns/groups · suggest hypotheses
│       (valuable on its own, not just prep work)
│
├── Diagnostic    → WHY did it happen?
│   ├── Goal: find causes, investigate causal relationships, suggest solutions
│   │   ⚠ Can suggest causes — cannot PROVE causality on its own
│   ├── Techniques: drill-down · correlation/regression · hypothesis testing · RCA
│   └── Root Cause Analysis (RCA): structured method to trace event to its source
│
├── Predictive    → WHAT will happen?
│   ├── Goal: anticipate outcomes, forecast processes, estimate unknowns
│   │   ⚠ Always a degree of uncertainty — predictions are not certainties
│   ├── Techniques:
│   │   ├── ML classification  (predicting subscription cancellations)
│   │   ├── ML regression      (predicting housing prices)
│   │   ├── Time series        (predicting sales revenue over time)
│   │   └── Predictive text    (spam detection)
│   └── Modeling: data split into training set + test set → evaluate with accuracy metric
│
└── Prescriptive  → WHAT should we do?
    ├── Goal: make informed decisions, optimize processes, mitigate risks
    ├── Techniques: rule-based systems · reinforcement learning · simulation · recommendation engine
    └── Recommendation engine: predicts interests from past behavior → provides recommendations
```

> **Decision rule:** Match the analytics type to the question — *What happened?* → Descriptive · *Why?* → Diagnostic · *What will?* → Predictive · *What to do?* → Prescriptive

### CHAPTER 4 — Communicating Insights

```
COMMUNICATING INSIGHTS
│
├── Data Visualizations
│   ├── Purpose: understand complex data at a glance, support EDA, dashboards, infographics
│   ├── Chart types:
│   │   ├── Trend          → line chart
│   │   ├── Relationships  → bar chart, scatter plot
│   │   └── Distributions  → histogram, box plot
│   └── What makes a good visualization:
│       ├── Eliminate clutter  → less is more, clarity and readability
│       ├── Focus attention    → link to insight, highlight key aspects
│       └── Provide context    → McCandless technique
│
│   McCandless technique (5 steps):
│   1. Introduce the visualization
│   2. Anticipate obvious questions
│   3. State the central insight
│   4. Provide supporting evidence
│   5. Closing statement
│
├── Data Storytelling
│   ├── Why: memorable · keeps attention · convinces audience · motivates action
│   ├── Components: Visualization + Narrative + Context (same as Course 2)
│   └── Narrative elements:
│       ├── Characters  → Stakeholders
│       ├── Problem     → Business problem
│       ├── Setting     → Relevant background
│       ├── Plot        → Central + supporting insights
│       └── Resolution  → Solution + recommendations
│
└── Three Keys to Communicating Effectively
    ├── Focus     → select right data · right visualizations · keep central message
    │   Central message = Problem + Insights + Impact (3-sentence formula)
    ├── Structure → storytelling + McCandless · visual layout · outline as blueprint
    │   Outline: Introduction (problem, context, objectives) → Body (data, analysis, findings)
    │            → Conclusions (insights, recommendations)
    └── Form      → medium (oral vs. written) · language (technical vs. non-technical)
                    scope (short vs. long-form) · audience (who, what they need, what they want)
```

## Chapter 1 — Data Literacy Basics

#### Chapter 1 — Lesson 1: Why data literacy is an essential skill *(Video / 50 XP)*

**🎯 GOAL — What is this lesson about?**
Data literacy is a life skill for the data era — not a technical skill. This lesson defines what data literacy means, establishes its three core components, and frames the entire course around them.

---

Data literacy is the ability to **read, work with, analyze, and communicate insights with data**. It has three components:

- **Reading data** — identifying relevant sources, collecting and managing data
- **Working with and analyzing data** — turning data into insights using analytics (descriptive, predictive, etc.)
- **Communicating insights** — demonstrating the value of insights and defining actions based on them; tools are visualizations and storytelling

Data literacy is not about coding or understanding the maths behind an analysis. It is about critically thinking about data — understanding its relevance, placing it in context, knowing both its potential and its limits. Like driving a car: you don't need to know what's under the hood, but you must know the difference between the brakes and the gas.

---

**💡 Why this matters**
A data analyst who cannot communicate insights produces findings that die in a spreadsheet. Data literacy defines the full chain from source to decision. Understanding the three-component structure helps any professional see where a breakdown happens — and what to fix: is the data being read correctly? Is the analysis appropriate? Is the insight reaching the decision-maker in a usable form?

**✅ RECAP**
- Data literacy = Read · Work with & Analyze · Communicate — three components, one chain
- It is a critical thinking skill, not a technical one — relevance, context, potential, limits
- No insight is valuable without communication — defining actions is as important as finding them

**🔗 Connection**
- The three components map directly onto the course structure: Ch2 = Reading · Ch3 = Analyzing · Ch4 = Communicating — this lesson is the thesis of the entire course
- The same three-part chain appears in Course 2 (*Communicating Data Insights*) — data literacy is the frame, storytelling is one execution of it

#### Chapter 1 — Lesson 5: From data to insights *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data-literacy/data-literacy-basics?ex=5) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
The DIKW pyramid (Data → Information → Knowledge → Wisdom) explains how raw observations become actionable understanding — and why *insights* are the realistic bridge between knowledge and wisdom.

---

The **DIKW pyramid** has four levels, each adding something new:

| Level | What it is | What is added | Weather example |
|-------|-----------|---------------|-----------------|
| **Data** | Raw observations, unorganized, no meaning yet | — | 15°C · dark clouds · raindrops |
| **Information** | Data placed into context, organized or aggregated | Context | Temperature dropped → clouds formed → rain started |
| **Knowledge** | Information combined to detect patterns and make connections | Meaning | Temp drop + dark clouds = rain coming |
| **Wisdom** | Applied knowledge — acting proactively | Action | I know when rain comes → I bring my umbrella |

The step from Knowledge to Wisdom is the hardest — wisdom implies complete, perfect knowledge, which is never fully reachable. **Insights** are the solution: realistic gold nuggets of applied knowledge that move us closer to wisdom step by step. Each new insight (humidity predicts better than temperature → atmospheric pressure predicts days ahead) improves the picture without requiring perfection.

---

**💡 Why this matters**
Every data project sits somewhere on the DIKW pyramid. Raw data from a database is not an insight — it is the starting point. A dashboard showing trends is information. A pattern identified across multiple dashboards is knowledge. A recommendation to a decision-maker — *do this because of what the data shows* — is wisdom. Knowing where your current work sits on the pyramid tells you what the next step is.

**→** *The goal is not perfect wisdom — it's accumulating enough insights to act better than before.*

**✅ RECAP**
- Data → Information: context is added (organize, aggregate)
- Information → Knowledge: meaning is added (patterns, connections, predictions)
- Knowledge → Wisdom: action is added — but perfect wisdom is unreachable
- Insights bridge the gap: realistic, achievable units of applied knowledge
- Each new insight brings us one step closer to the full picture

**🔗 Connection**
- The DIKW pyramid is the theoretical spine of the entire course — Ch2 (reading) operates at the Data/Information level, Ch3 (analyzing) at the Knowledge level, Ch4 (communicating) at the Wisdom/Insight level
- "Insights" as the output of analysis is the same concept Course 2 built around — storytelling exists to communicate insights, not raw data


#### Chapter 1 — Lesson 6: DIKW pyramid as a process *(Exercise / 100 XP)*

---

**🎯 GOAL — What does this exercise test?**
Can you translate the four DIKW levels from abstract labels into concrete actions — matching each step to what you actually *do* at that level.

---

**📋 Scenario**
Getting from data to wisdom according to the DIKW framework takes different steps, each taking us one level higher on the pyramid.

**Task:** Order the items to match the DIKW steps: Data · Information · Knowledge · Wisdom.

---

**✅ Solution**

| Order | DIKW Level | Action |
|-------|-----------|--------|
| 1 | **Data** | Make observations and measurements |
| 2 | **Information** | Put raw data into context |
| 3 | **Knowledge** | Connect different pieces of information and gain meaning |
| 4 | **Wisdom** | Apply knowledge to take action |

> DataCamp: *"Gaining meaning is the bridge between information and knowledge."* — the most commonly confused step. Meaning is added at the Knowledge level, not the Information level.

---

**💡 Why this matters**
The common failure in data work is stopping at Information — producing a dashboard, a report, a chart — and calling it done. The DIKW framework makes the gap visible: organized data is not insight, and insight is not action. Knowing which step you are currently on tells you exactly what is still missing before a decision can be made.

**→** *If your deliverable doesn't change how someone acts, you haven't reached Wisdom yet — keep going.*

**✅ RECAP**
- Data = observations and measurements (raw, no meaning)
- Information = data in context (organized, aggregated)
- Knowledge = meaning extracted (patterns, connections) — *gaining meaning* is the bridge from Information
- Wisdom = knowledge in action (proactive, decision-enabling)
- The most common trap: confusing Information (organized data) with Knowledge (meaning extracted)

**🔗 Connection**
- This exercise operationalises L5 — the pyramid is no longer a diagram, it's a checklist of what you still need to do
- "Apply knowledge to take action" is the definition of the Communicating component from L1 — DIKW and the three data literacy components describe the same journey from two angles


#### Chapter 1 — Lesson 7: What is insight? *(Exercise / 50 XP)*

---

**🎯 GOAL — What does this exercise test?**
Can you distinguish the precise role of *insight* in the DIKW pyramid — separating what insight does from what wisdom does, since they are easily confused.

---

**📋 Scenario**
Insight is the bridge between knowledge and wisdom.

**Task:** What is **not** a characteristic of insight?

---

**✅ Solution**

| Option | Characteristic of insight? | Why |
|--------|---------------------------|-----|
| Insight allows us to get closer to the ideal of wisdom | ✅ Yes | Insights accumulate and bring us step-by-step closer to wisdom |
| Insight is valuable, and can realistically be achieved | ✅ Yes | Insight is described as the "gold nugget" — achievable, unlike perfect wisdom |
| **Insight allows us to apply knowledge and take action** | ❌ No — correct answer | Applying knowledge to take action is the definition of *wisdom*, not insight. Insight bridges knowledge and wisdom — the action belongs to wisdom |
| Wisdom can be approached, but not quite fully reached | ✅ Yes | This is also a property of insight's role — insight is how we approach wisdom without claiming to have reached it |

> DataCamp: *"Insight is the bridge between knowledge and wisdom."* — The trap: option 4 describes the nature of wisdom (unreachable ideal), which is also where insight leads. Option 3 describes the action that belongs to wisdom alone.

---

**💡 Why this matters**
In practice, the insight/wisdom boundary matters when presenting findings. Presenting an insight is saying: *"here is what the data shows."* Wisdom is the step after: *"here is what we should do."* A data analyst produces insights — the decision-maker applies them as wisdom. Confusing the two either oversteps the analyst's role or leaves the decision-maker without direction.

**✅ RECAP**
- Insight = bridge between knowledge and wisdom — valuable, realistic, accumulative
- Wisdom = applying knowledge to take action — the step insight enables but does not itself perform
- The trap: "applying knowledge to take action" sounds like insight but is the definition of wisdom
- Each new insight moves us closer to wisdom; wisdom itself is the destination, never fully reached

#### Chapter 1 — Lesson 9: Data-driven decision making *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data-literacy/data-literacy-basics?ex=9) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Data-driven decision making is not a technology — it is a five-step process anchored by a well-crafted problem statement. This lesson defines the process, dismantles three common misconceptions, and shows what a good problem statement looks like in practice.

---

**Data-driven decision making** is the process of using data to make an informed decision about a specific problem and acting upon it. It serves four purposes: optimising performance, gaining a better understanding of a situation, protecting against risks, and determining the best course of action.

**Three misconceptions — all wrong:**

| Misconception | Why it is wrong |
|---------------|----------------|
| Only for large organizations | You can start small — any organization with a problem and data can use it |
| Only the data team's responsibility | Including people with different backgrounds enriches the process |
| Always the best option | Data-driven decision making is a strategy — it has advantages and disadvantages; it is not always appropriate |

**The 5-step data-driven process:**

| Step | Action |
|------|--------|
| 1 | Define a problem statement |
| 2 | Collect the necessary data |
| 3 | Perform data analysis |
| 4 | Communicate the results |
| 5 | Take action and reflect |

**The problem statement** is Step 1 — and the most important. It answers: *"What is the problem you want to solve?"* It guides which data to collect, which analysis technique to use, and how results can be acted on. A good problem statement is:
- **Clearly defined** — specific, not vague
- **Actionable** — a specific action is possible based on the answer
- **Realistic** — solvable with available resources

Three guiding questions to craft one: *What is the current situation? What do we need to know? Where do we want to be?*

**Two examples:**

| Context | Starting question | Problem statement |
|---------|------------------|-------------------|
| Customer service team | How can we improve customer satisfaction? | Customer satisfaction is currently under our target. We need to know the main reasons for a negative score so we can take measures to improve it. |
| University | Why are average math scores declining for three years? | Math scores have declined for three years. We need to know the main causes so we can take measures to counter them. |

---

**💡 Why this matters**
Most data projects fail not at the analysis stage but at the framing stage — a vague or unmeasurable problem leads to analysis that produces no actionable output. The five-step process makes the sequence explicit: without a good problem statement in Step 1, Steps 2–5 have no direction. In practice, a data analyst who pushes back on a poorly framed request (*"we need to know everything about sales"*) and helps define a crisp problem statement (*"we need to identify the top three reasons for cart abandonment in Q2"*) delivers more value than one who starts querying immediately.

**→** *Step 1 is not preparation — it is the work. A precise problem statement is already half the solution.*

**✅ RECAP**
- Data-driven decision making = using data to make an informed decision and act on it
- Three misconceptions: not only for large orgs · not only the data team · not always the best option
- 5-step process: Define → Collect → Analyse → Communicate → Act & Reflect
- Problem statement: the anchor of the entire process — clearly defined · actionable · realistic
- Three problem categories: describe state · diagnose causes · detect anomalies or predict

**🔗 Connection**
- The 5-step process maps directly onto the DIKW pyramid: Define problem (Data level) → Collect & Analyse (Information/Knowledge) → Communicate & Act (Wisdom/Insight)
- Problem statement categories (describe · diagnose · detect/predict) reappear in Ch3 as the four analytics types — descriptive, diagnostic, predictive, prescriptive are the analytical answers to exactly these categories

---

**📋 Scenario**
Some analysis has been carried out to compare test scores in mathematics, reading, and writing between two groups: students who completed the test preparation course and students who did not.

Three charts (one per subject) each show: x-axis = whether students took the course ("none" vs. "completed"); y-axis = test scores 0–100; a blue trend line connecting the average score of both groups; individual scores shown as dots (❌ = no course, ✅ = completed course). **✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| Do students who score high on one test also score higher on the other final tests? | ❌ | This asks about correlation *between* subjects — not about the effect of the preparation course. The analysis groups by course completion, not by cross-subject performance |
| Which students are more likely to ace their classes? | ❌ | Too vague — "ace their classes" is undefined and not measurable. A good problem statement is specific and actionable, not a general aspiration |
| **Does completion of the test preparation course improve student performance on the final tests?** | ✅ Correct | This matches the analysis exactly: the grouping variable is course completion; the outcome variable is test scores; three subjects are compared — *"Clearly defining a problem beforehand goes a long way toward successful data-driven decision making"* |
| Are there gender differences between the test scores? | ❌ | The charts show no gender variable — the x-axis is course completion, not gender. This question would require a completely different dataset |

> DataCamp: *"Great! Clearly defining a problem beforehand goes a long way toward successful data-driven decision making."*

**💡 Why this matters**
Reading an analysis backwards — from charts to the question that produced them — is a real skill. In practice, you will often receive dashboards or reports with no explicit problem statement attached. The data does not contain a gender variable — so that question would require re-collecting data, not re-reading these charts.

**✅ RECAP**
- A good problem statement is specific, actionable, and measurable — "which students are more likely to ace their classes" fails all three
- Read the grouping variable in the chart: x-axis = course completion → the question is about course completion, not correlations or gender
- Reverse-engineering the problem statement from an analysis is a real analytical skill — not just a DataCamp exercise
- Option 4 is the most dangerous distractor: a real question, but not the one this data can answer

### 🧠 Chapter 1 Recap — Data Literacy Basics

*13 lessons · 800 XP · Arc: what data literacy is → how it works (DIKW) → how it drives decisions*

---

**The central question Chapter 1 answers:**
> What does it mean to be data-literate — and how do you get from raw data to a decision that matters?

---

**Three-part arc:**

**Part 1 — What data literacy is (L1–L4)**

Data literacy is the ability to read, work with, analyze, and communicate data — critically engaging with data without needing technical expertise. It has three components: Reading (databases), Analyzing (descriptive and predictive analytics), and Communicating (storytelling and visualizations). It applies to every profession — nurse, teacher, HR specialist, marketing manager. Technical skill is not a requirement; critical engagement with data is.

**Part 2 — How it works: the DIKW pyramid (L5–L8)**

The DIKW pyramid is the engine underneath data literacy. Data (raw observations) → Information (data in context) → Knowledge (meaning extracted, patterns identified) → Wisdom (knowledge applied to action). Wisdom is the unreachable ideal; **Insights** are the realistic gold nuggets that bridge Knowledge and Wisdom — each new insight moves us one step closer without claiming to have arrived. The most commonly confused transition: Information → Knowledge requires extracting *meaning* (patterns, predictions), not just organizing data — that confusion is tested directly in L6 and L8.

**Part 3 — How it drives decisions: data-driven decision making (L9–L13)**

DDDM is the five-step process that operationalises the DIKW pyramid: Define → Collect → Analyze → Communicate → Act & Reflect. Step 1 (Define) is non-negotiable — a clear, actionable, realistic problem statement determines everything that follows. Three misconceptions consistently tested: DDDM is not only for large organisations · not only the data team's responsibility · not always the best strategy (no right data = difficult to impossible). The process is domain-independent: retail, healthcare, education, manufacturing all follow the same structure.

---

**The key principle that runs through all 13 lessons:**

> *Clarity before collection. Problem before data. Insight before action.*

Every lesson in Ch1 is a variation of this principle. The DIKW pyramid says: don't confuse organized data with meaning. The problem statement rule says: don't collect data before defining the question. L13 says: don't inherit someone else's analysis without first identifying what question it was designed to answer.
