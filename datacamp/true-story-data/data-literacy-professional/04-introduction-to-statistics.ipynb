### CHAPTER 1 — What is Statistics?

```
WHAT IS STATISTICS?
│
├── The Field of Statistics
│   └── Practice and study of collecting and analyzing data
│
├── Two Main Branches
│   ├── Descriptive / Summary Statistics
│   │   └── Describing or summarizing data (e.g. total crime per borough)
│   └── Inferential Statistics
│       └── Collect a sample → apply results to the population
│           (e.g. how many purchase clothing after social media ads?)
│
├── What Statistics Can Do
│   ├── Answer practical questions (avg salary, customer inquiries per week)
│   ├── Applications across society (safer products, government policy)
│   └── Validate scientific breakthroughs (Covid-19 vaccines)
│
├── Limitations
│   ├── Requires specific, measurable questions
│   └── Cannot find out WHY relationships exist
│
├── Types of Data: Numeric
│   ├── Continuous data (stock prices: $2328.14, $156.77)
│   └── Interval / Count data (cups of coffee per day: 4, 2, 3)
│
├── Types of Data: Categorical
│   ├── Nominal data (eye color: brown, green, blue)
│   └── Ordinal data (agreement scale: strongly agree → strongly disagree)
│
├── Visualizing Data
│   ├── Numeric data → histograms
│   └── Categorical data → bar charts
│
├── Measures of Center
│   ├── Mean  → sum of all values / count (sensitive to outliers)
│   ├── Median → middle value when sorted (robust to outliers)
│   └── Mode  → most frequent value
│
└── Measures of Spread
    ├── Range  → max − min
    ├── Variance → avg squared distance from mean
    ├── Standard Deviation → √variance (close to 0 = clustered around mean)
    ├── Quartiles → split data into 4 equal parts (Q1, Q2=median, Q3)
    ├── IQR → Q3 − Q1 (less affected by extreme values)
    └── Box plots → visual summary using quartiles
```

> **Running example:** London crime data (Metropolitan Police Service) — burglary, robbery, theft, vehicle crime across 32 boroughs

### CHAPTER 2 — What are the Chances?

```
PROBABILITY
│
├── Measuring Chance
│   └── P(event) = (# ways event can happen) / (total # possible outcomes)
│       Example: P(heads) = 1/2 = 50%
│
├── Sampling
│   ├── With replacement → each pick is independent, probabilities stay constant
│   └── Without replacement → each pick is dependent, probabilities change
│
├── Independent Events
│   └── Probability of second event does NOT change based on first event
│
├── Dependent Events
│   └── Probability of second event IS affected by outcome of first event
│
├── Conditional Probability
│   ├── P(A|B) = P(A ∩ B) / P(B)
│   └── Used to calculate probability of dependent events
│
├── Venn Diagrams
│   └── Visual representation of event overlaps and conditional probability
│
├── Discrete Probability Distributions
│   ├── Describe probabilities for discrete outcomes (e.g. die roll)
│   ├── Expected value = mean of a probability distribution
│   ├── Discrete uniform distribution (fair die: each outcome = 1/6)
│   └── Probability = area under the distribution
│
├── Law of Large Numbers
│   └── As sample size increases, sample mean approaches expected value
│
└── Continuous Distributions
    ├── Continuous uniform distribution (e.g. bus wait time 0–12 min)
    ├── Total area under any probability distribution = 1
    ├── Bimodal distribution (two peaks)
    └── Normal distribution (bell curve, symmetrical)
```

> **Running example:** Online retail sales data — product types (Basket, Art & Sculpture, Jewelry, Kitchen, Home Decor) with order counts and probabilities

### CHAPTER 3 — Key Probability Distributions

```
KEY DISTRIBUTIONS
│
├── Binomial Distribution
│   ├── Probability of # successes in a sequence of independent events
│   ├── Described by n (total events) and p (probability of success)
│   ├── Expected value = n × p
│   ├── Requires independence — if events are dependent, binomial does NOT apply
│   └── Applications: clinical trials (effective/not), sports betting (win/lose)
│
├── Normal Distribution
│   ├── Symmetrical bell curve, total area = 1, curve never hits 0
│   ├── Described by mean (center) and standard deviation (spread)
│   ├── 68-95-99.7 Rule:
│   │   ├── 68% within 1 std dev
│   │   ├── 95% within 2 std dev
│   │   └── 99.7% within 3 std dev
│   ├── Many real-world datasets resemble normal distribution
│   ├── Required for many statistical tests
│   ├── Skewness → asymmetry (left or right tail)
│   └── Kurtosis → occurrence of extreme values (leptokurtic, mesokurtic, platykurtic)
│
├── Central Limit Theorem (CLT)
│   ├── Sampling distribution of a statistic approaches normal as sample size increases
│   ├── Applies to both means and proportions
│   ├── Samples must be random and independent
│   └── Why it matters: enables inference without knowing population distribution
│
└── Poisson Distribution
    ├── Probability of # events occurring over a fixed period of time
    ├── λ (lambda) = average # events per time period = expected value
    ├── λ is the distribution's peak
    ├── Poisson processes: known avg rate, random timing between events
    │   (animal adoptions/week, restaurant arrivals/hour, website visits/day)
    └── CLT still applies to the Poisson distribution
```

> **Running example:** Coin flips (binary outcomes, binomial), die rolls (CLT demonstration), bus wait times (Poisson processes)

### CHAPTER 4 — Hypothesis Testing and Correlation

```
HYPOTHESIS TESTING
│
├── Why Hypothesis Testing?
│   ├── Used to compare populations
│   ├── Applications: price change → revenue, website change → traffic, medication → treatment
│   └── Dates back to 1700s (human sex ratio)
│
├── Null and Alternative Hypotheses
│   ├── Null (H₀): assume no difference exists
│   └── Alternative (H₁): a difference exists
│
├── Hypothesis Testing Workflow
│   ├── Define target populations
│   ├── Develop null and alternative hypotheses
│   ├── Collect or access sample data
│   ├── Perform statistical tests
│   └── Draw conclusions about the population
│
├── Sample Size
│   ├── CLT: larger samples → means closer to population mean
│   └── Trade-off: time and resource intensive
│
├── Variables
│   ├── Independent variable → unaffected by other data (e.g. vitamin C supplementation)
│   └── Dependent variable → affected by other data (e.g. birth gender ratio)
│
├── Experiments
│   ├── Treatment = independent variable, Response = dependent variable
│   ├── Controlled experiments: treatment group vs. control group
│   ├── Randomization → participants assigned randomly (RCT)
│   ├── Blinding → participants don't know which group
│   ├── Double-blind → administrator also doesn't know
│   └── RCT vs. A/B testing: RCT = multiple groups, science; A/B = two groups, marketing/engineering
│
├── Correlation
│   ├── Pearson correlation coefficient (r) → -1 to +1
│   ├── Magnitude = strength (0.99 very strong, 0.75 strong, 0.56 moderate, 0.21 weak, 0.04 none)
│   ├── Sign = direction (+ = same direction, - = opposite)
│   ├── Linear relationships only
│   ├── Correlation ≠ causation
│   └── Confounding variables → unmeasured factor affecting both variables
│
└── Interpreting Results
    ├── p-value → probability of achieving this result assuming H₀ is true
    ├── Significance level (α) → threshold for rejecting H₀ (typical: 0.05)
    ├── If p ≤ α → reject H₀ (statistically significant)
    ├── Type I Error → false positive (reject H₀ when H₀ is true)
    └── Type II Error → false negative (fail to reject H₀ when H₀ is false)
```

> **Running example:** Vitamin C supplementation and birth gender ratio (Ch4), Chicago vs. Bangkok life expectancy (p-value demonstration), Gym costs vs. water costs (correlation ≠ causation)

**🎯 GOAL — What is this lesson about?**
Understand the definition of statistics, its two main branches (descriptive and inferential), and the types of practical questions statistics can and cannot answer.

---

**Statistics** is the practice and study of collecting and analyzing data. It splits into two main branches:

| Branch | Definition | Real-world example |
|--------|------------|-------------------|
| **Descriptive / Summary Statistics** | Tools used to describe or summarize data | Thefts in Westminster account for 36% of all thefts in five London boroughs |
| **Inferential Statistics** | Using a sample to draw conclusions about a population | Survey 100 people on social media advertising → infer behavior for all people |

Statistics is everywhere — from sports announcers discussing player performance to personal finance tracking. It allows us to answer specific, measurable questions: *What is the average USA salary?* *How many customer inquiries will a company receive per week?* Applications span product safety, government policy, and scientific validation (e.g., Covid-19 vaccines were 89% effective in preventing severe disease in older adults in the UK).

**Important limitation:** Statistics requires specific, measurable questions. It can answer *whether* rock music outsells jazz (measurable), but **not** *why* people prefer one genre over another (open-ended). Similarly, statistics reveals patterns (women live longer than men) but cannot explain the underlying reasons.

**Data types matter.** To analyze data correctly, we must first classify it:
- **Numeric (Quantitative):** Continuous (stock price) or count/interval (cups of coffee per day)
- **Categorical (Qualitative):** Nominal, unordered (eye color) or ordinal, ordered (survey: strongly disagree → strongly agree)

Visualization choices follow: numeric data pairs appear as scatter plots; categorical-numeric relationships appear as grouped bar charts (e.g., thefts by London borough).

---

**💡 Why this matters**
In any analytics role, you will inherit questions before you inherit data. Statistics gives you the language to know which questions are answerable and which require a different approach. A recruiter will ask: *"Why didn't you use statistics for that?"* Understanding the boundary — measurable vs. philosophical — separates analysts who understand their tools from those who misapply them. It is also the foundation for every concept that follows in this course.

**✅ RECAP**
- Statistics = practice of collecting and analyzing data to answer specific, measurable questions
- Two branches: **Descriptive** (summarize data) and **Inferential** (sample → population conclusion)
- Data types: **Numeric** (continuous/count) and **Categorical** (nominal/ordinal)
- Statistics cannot answer *why* — only *what* and *whether*
- Applications: product safety, policy, scientific validation; limitations: requires measurable questions

**🔗 Connection**
- Descriptive statistics (this lesson's branch 1) is the focus of Chapter 1 — we will learn the tools (mean, median, standard deviation)
- Inferential statistics (this lesson's branch 2) connects to Chapters 2–4 — probability, distributions, hypothesis testing
- Data types classification (this lesson) determines which visualization and summary statistic to choose — practiced in Lesson 2 onwards

The skill you're practicing here is knowing when to say "we need to reframe this question" versus "let's build the analysis." This is a core communication skill for analysts and data engineers.

**→** *When a question feels unanswerable, your job is to translate it into something measurable — not to build analysis on an impossible foundation.*

**✅ RECAP**
- Statistics answers specific, measurable questions only
- A/B testing (new vs. current) is a classic statistics use case — inferential design
- "Why" questions require qualitative research or domain expertise, not statistics
- Randomness cannot be predicted (lottery, stock prices tomorrow)
- Your first skill: recognizing which questions statistics can solve before investing time

A recruiter will notice if you treat ordinal like continuous in your first project.

**→** *Data type = decision gate. Get it wrong early, and every analysis that follows is built on the wrong foundation.*

**✅ RECAP**
- **Continuous data:** measured on a continuous scale; infinite precision (lap times, heights, prices)
- **Nominal data:** unordered categories; no inherent ranking (positions, colors, employment status)
- **Ordinal data:** ordered categories; clear ranking but no measurable interval between ranks (satisfaction levels, income brackets)
- Data type determines visualization and summary statistic choice (next lessons)
- Confusing types = wasted analysis effort; recognizing types early = correct decisions downstream

#### Chapter 1 — Lesson 4: Descriptive vs. Inferential statistics *(Exercise / 100 XP)*

---

**🎯 GOAL — What does this exercise test?**
Distinguish between descriptive statistics (summarizing existing data) and inferential statistics (using a sample to draw conclusions about a population) by analyzing real-world scenarios.

---

**📋 Scenario**

Recall that there are two main branches of statistics — descriptive statistics and inferential statistics.

Understanding what type of statistics is required for a given situation is an essential skill in order to draw accurate conclusions.

Now you have the opportunity to check your knowledge of how these two branches of statistics are used in practice. You'll classify each scenario as a task for descriptive or inferential statistics.

**Task:** Choose which type of statistics to use for each scenario. Map each item into the Descriptive or Inferential bucket by dragging.

---

**✅ Solution**

| Scenario | Correct Type | Classification | Why |
|----------|--------------|-----------------|-----|
| Given data on all projects completed, what percentage were delivered on time? | Descriptive | Summarization | You have the complete dataset (all projects). Calculate a summary statistic (percentage) to describe what already exists — no inference needed |
| Given data on all purchases, what is the average amount spent per customer? | Descriptive | Summarization | Complete dataset provided (all purchases). Calculate the mean to summarize existing data — a purely descriptive task |
| After surveying 100 adults on how many hours of sleep they get per night, what percentage of *all adults* sleep for at least 6 hours per night? | Inferential | Sample → Population | Only a sample of 100 adults surveyed, not all adults. Use this sample to **infer** what percentage of the *entire population* of adults meets the criterion — classic inferential logic |
| Given test scores for 50 students at one high school, what is the average score for *all students in that school*? | Inferential | Sample → Population | Only 50 students' data — a sample of the school. Use this sample to **infer** the average for all students in the school (the population) |

**Common mistake:** Confusing "all students *in that school*" (population to infer about) with "all students in the sample" (descriptive). If you have scores for *all* students in the school, that is descriptive. If you have 50 students and need to infer for the remaining ~500, that is inferential.

---

**💡 Why this matters**

Recognizing the type of statistics required to answer a given question allows you to choose the appropriate method and, by extension, produce the most accurate answer. A data analyst who uses descriptive analysis when inferential analysis is needed will draw incorrect conclusions — *"We surveyed 50 customers and found they spent an average of $50; therefore all customers spend $50 on average"* is flawed. By contrast, using inferential methods correctly (confidence intervals, hypothesis tests) lets you to say with quantified certainty: *"We are 95% confident the true average is between $48 and $52."* This precision is the difference between a weak analysis and one a recruiter will respect.

**→** *Type mismatch = credibility loss. Get the method right first, then execute it.*

**✅ RECAP**
- **Descriptive statistics:** summarize data you already have (percentages, averages, distributions of complete datasets)
- **Inferential statistics:** use a sample to draw conclusions about a larger population (hypothesis tests, confidence intervals, predictions)
- Complete dataset → descriptive; sample of a larger population → inferential
- The wording "for all [population]" signals inferential; "what is the [statistic] of this dataset" signals descriptive
- Mixing these up undermines credibility — method choice is step 1 before any analysis

#### Chapter 1 — Lesson 5: Measures of center *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-statistics/summary-statistics-ef7cd595-e629-4103-a4ad-152b9e13b909?ex=5) *(requires login)*

**🎯 GOAL — What is this lesson about?**
Learn the three main measures of center (mean, median, mode), understand how to calculate each, and recognize which measure is most appropriate depending on data distribution and presence of outliers.

---

**Measures of center** are statistics that describe the "typical" or central value of a dataset. They appear constantly in daily conversation: *average salary, typical house cost, most common hair color*.

| Measure | Definition | Formula | When to use | Strength | Weakness |
|---------|-----------|---------|-------------|----------|----------|
| **Mean** | Average; sum of all values divided by count | Σx / n | Symmetric data, no outliers | Uses all data points; mathematically convenient | Pulled toward outliers; can misrepresent skewed data |
| **Median** | Middle value when data is sorted; 50% below, 50% above | Middle value (or average of two middle values if n is even) | Skewed data, presence of outliers | Robust to outliers; unaffected by extreme values | Ignores magnitude of values; less mathematically convenient |
| **Mode** | Most frequent value | Count occurrences | Categorical data; finding the most common category | Only measure suitable for purely categorical data | Not useful for continuous data; may not exist or be non-unique |

**Example — London crime dataset (32 boroughs, 2 years):**
- **Mean burglaries per borough:** ~3,463 (sum all burglaries ÷ 32 boroughs)
- **Median burglaries per borough:** 3,416.5 (average of 16th and 17th values when sorted — slightly lower than mean)
- **Mode crime type:** Theft (most frequently occurring crime across all boroughs)

**Data distribution matters.** A symmetric histogram (equal tails) means mean and median are close — either works. A skewed histogram (outliers pulling one tail) means the mean gets pulled toward the extreme values, while the median stays centered. Example: robbery data with one borough having an unusually high count — that outlier inflates the mean but barely affects the median.

**→** *Outliers are the test: if the mean and median differ significantly, use the median.*

---

**💡 Why this matters**
In your first analytics project, a stakeholder will ask: *"What's the typical value?"* Your answer determines whether the company invests in a strategy. If you report the mean of skewed data (salary distribution, customer spend), you'll overstate the typical case. Reporting the median protects you — it reflects what 50% of the data actually experiences. A recruiter will notice if your summary statistics are pulled by outliers. Learning to recognize when mean vs. median is appropriate is the difference between a careless analyst and one who thinks critically about data.

**→** *Always check the distribution before choosing your measure. A single number without context is a liability.*

**✅ RECAP**
- **Mean:** sum ÷ count; works well for symmetric data; sensitive to outliers
- **Median:** middle value; robust to outliers; best for skewed data
- **Mode:** most frequent value; only suitable for categorical data
- Histograms reveal distribution shape — symmetric vs. skewed determines which measure to use
- Outliers pull the mean away from the median — when they differ significantly, prefer the median

#### Chapter 1 — Lesson 9: Measures of spread *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-statistics/summary-statistics-ef7cd595-e629-4103-a4ad-152b9e13b909?ex=9) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Understand measures of spread (range, variance, standard deviation, quartiles, IQR, box plots) — which describe how far apart data points are and quantify the variability in a dataset.

---

**Spread** describes how far apart data points are from each other. While measures of center (mean, median, mode) tell you the "typical" value, measures of spread tell you whether that typical value is representative of the whole dataset or if there's huge variability.

**Example:** T-shirts typically cost $30. But if prices range from $10 to $200, how likely are you to find one at $30? If prices range from $20 to $50, the typical price means much more.

| Measure | Definition | Formula / Calculation | When to use | Strength | Weakness |
|---------|-----------|----------------------|-------------|----------|----------|
| **Range** | Difference between max and min values | max − min | Quick spread estimate; simple to understand | Easy to calculate; intuitive | Ignores distribution shape; heavily affected by outliers |
| **Variance** | Average squared distance from each point to the mean | Σ(x − mean)² / n | Theoretical measure for statistical testing | Uses all data points; mathematical convenience | In squared units (hard to interpret); sensitive to outliers |
| **Standard Deviation (SD)** | Square root of variance; typical distance from mean | √variance | Most common spread measure; same units as data | Interpretable in original units; mathematically powerful | Sensitive to outliers; harder to compute manually |
| **Quartiles** | Values dividing data into 4 equal parts (Q1, Q2, Q3, Q4) | 25th, 50th, 75th, 100th percentiles | Understanding distribution shape; robust alternative | Unaffected by outliers; easy to understand | Loses fine-grained information |
| **IQR (Interquartile Range)** | Distance between Q1 and Q3; middle 50% spread | Q3 − Q1 | Robust spread measure; identifies outliers | Outlier-resistant; focuses on central bulk | Ignores extreme values entirely |
| **Box Plot** | Visual: box (Q1 to Q3) + median line + whiskers | Q1, median, Q3, ± 1.5×IQR for whiskers | Visualizing distribution shape and outliers | Compact visual summary; shows skewness | Limited precision; space-intensive for many groups |

**Example — London crime dataset:**

**Burglaries:**
- Range: 5,183 (Tower Hamlets) − 1,432 (Kingston upon Thames) = 3,751
- Variance: ~233 million (burglary count squared)
- Standard Deviation: ~15,241 (average distance from mean of ~3,463)
- Q1: ~2,500; Median (Q2): ~3,416.5; Q3: ~4,392
- IQR: 4,392 − 2,500 = 1,892 (middle 50% of boroughs differ by ~1,892 burglaries)

**Vehicle Crimes:**
- Histogram shows narrower spread than burglaries — values clustered closer to mean
- Same analysis structure; different numeric results

**Decision tree for choosing a spread measure:**
- **Quick estimate?** → Range
- **Comparing to mean/SD?** → Variance or Standard Deviation
- **Comparing distributions visually?** → Box plot
- **Resisting outliers?** → IQR or quartiles
- **Formal statistical testing?** → Standard Deviation (paired with confidence intervals, hypothesis tests)

---

**💡 Why this matters**
A product manager asks: *"Is our customer spend consistent or all over the map?"* If mean spend is $100 but SD is $5, customers are predictable. If SD is $200, they're unpredictable — you need different strategies. A financial analyst reporting on portfolio returns always pairs mean with SD (or variance) — omitting spread is incomplete and misleading. A recruiter will notice if you report center without spread. Together, mean + SD (or median + IQR) paint the full picture of your data.

**→** *Center without spread is half a story. Always report both.*

**✅ RECAP**
- **Spread = variability:** how far apart data points are
- **Range:** max − min; simple but outlier-sensitive
- **Variance:** average squared distance from mean; mathematical convenience; hard to interpret (squared units)
- **Standard Deviation:** √variance; converts back to original units; most interpretable and commonly reported
- **Quartiles & IQR:** split data into 4 parts; IQR (Q3 − Q1) is outlier-resistant
- **Box plot:** visual summary of quartiles, median, and outliers
- Center + Spread = complete data summary; never report one without the other

**🎯 GOAL — What does this exercise test?**
Compare standard deviations across two distributions by examining histogram x-axis range and spread — understanding that wider x-axis range indicates larger SD.

---

**📋 Scenario**

For the final exercise of the chapter, it's time to see how you interpret spread using the standard deviation.

Two histograms are displayed: one for Public Order Offences and another for Miscellaneous Crimes Against Society.

Pay close attention to the **x-axis scale on each histogram**. Since the standard deviation is measured in the same units as the data, the range and spread of values along the x-axis is key to determining which distribution has a larger standard deviation.

**Task:** Which type of crime has a larger standard deviation?

**Possible answers:**
1. It's not possible to determine which has the larger standard deviation based on the plots
2. Miscellaneous Crimes Against Society
3. Public Order Offences ✅ **CORRECT** (and your choice)

---

**✅ Solution & Walkthrough**

**Correct answer: Public Order Offences (Option 3)**

**Reading the x-axis scales:**

| Crime Type | X-axis min | X-axis max | Range | Spread pattern |
|-----------|-----------|-----------|-------|-----------------|
| **Public Order Offences** | ~1000 | ~6500 | ~5500 units | Spread across wider range; histogram stretched horizontally |
| **Miscellaneous Crimes Against Society** | ~200 | ~1250 | ~1050 units | Compressed range; histogram tightly clustered |

**Standard deviation interpretation:**
- **Public Order Offences:** SD ≈ 1000–1500 (rough estimate from histogram width)
- **Miscellaneous Crimes:** SD ≈ 200–400 (rough estimate — much tighter)
- **Conclusion:** Public Order Offences has **larger spread = larger SD**

**Why this works:** Histograms with wider x-axis ranges (data spread over larger values) indicate larger standard deviations. The data in Public Order Offences is scattered across ~5500 units; Miscellaneous Crimes is packed into ~1050 units. Wider spread → larger SD.

---

**Why Option 2 fails:**

| Option | Answer | Verdict | Why |

| 1 | "It's not possible to determine..." | ❌ Incorrect | We CAN determine SD from histograms by reading x-axis spread. This is not a trick question. |

---

**Reading strategy for comparing SDs from histograms:**
1. **Look at x-axis scale min and max** for each histogram
2. **Calculate or estimate the range** (max − min)
3. **Wider range = larger SD** (data points further from mean, on average)
4. **Ignore y-axis height** — that's frequency, not spread
5. **Ignore histogram shape** (normal, skewed, etc.) — SD is about spread, not shape

**Common mistake:**
Confusing y-axis frequency with spread. A tall histogram (many data points in a bin) doesn't mean large SD. A histogram stretched wide on the x-axis means large SD. You might see Public Order Offences and think "the peak is lower, so smaller SD" — but that's looking at **height** (frequency), not **width** (spread). **Always compare x-axis scales, never y-axis.**

**✅ RECAP**
- **Standard Deviation:** typical distance from mean; measured in same units as data
- **Reading SD from histograms:** wider x-axis range = larger SD
- **X-axis vs. Y-axis:** x-axis shows spread (SD); y-axis shows frequency (not relevant for SD comparison)
- **Histogram shape doesn't determine SD:** tall spike or flat distribution — SD is about x-axis width, not shape
