#### Chapter 1 — Lesson 1: Data basics *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-basics?ex=1) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Establish what data is, where it comes from, and how to classify it — the conceptual foundation before any analysis begins.

---

**📋 Key concepts covered:**

**1. Data is everywhere**
You encounter data constantly without noticing it. You *are* data yourself — your name, age, hobbies, number of shirts you own, cups of coffee you drink. Social media comments, videos, photos — all data. From the number of trees in a city to the distance between Earth and the Moon.

**2. What is data?**
"Data" comes from the Latin *datum* — meaning "fact." Facts are incredibly valuable, but only if we know their meaning. Raw facts without meaning are just noise.

**3. Data context**
Data gains value with context. Example: a graph of goals scored per season by Messi (football) and Ovechkin (ice hockey). Without knowing the rules of each sport, the comparison is meaningless.

> *Context includes details such as when or where data was collected — often called **metadata**. Context is not optional. It is what makes data useful.*

**4. Types of data — two dimensions:**

| Dimension | Type A | Type B |
|-----------|--------|--------|
| **Structure** | Structured — organised, labelled (spreadsheets, tables) | Unstructured — no labels, no order (videos, photos, interviews) |
| **Nature** | Quantitative — numerical (goals scored, height, temperature) | Qualitative — categorical (favourite sport, customer feedback) |

Neither is better. The right type depends on the question being asked.

---

**✅ RECAP — What to remember**
- Data = facts. Facts need context to have meaning.
- Two classification axes: structure (structured/unstructured) and nature (quantitative/qualitative).
- Metadata = the context around data (when, where, how it was collected).
- The most common mistake: treating data as self-explanatory. It never is.

#### Chapter 1 — Lesson 2: Structured vs. unstructured *(Exercise / 100 XP)*

---

**📋 Scenario**
Emma just started her new job at DataPrint, an online store for printer gear. In her first week, she discovered loads of data and needs to sort it into **structured** or **unstructured** categories.

**Task:** Classify each item as structured or unstructured data.

---

**✅ Solution**

| Item | Classification | Why |
|------|---------------|-----|
| Customer information (Name, Age, ID) | **Structured** | Organised in rows and columns, predefined format |
| Financial transactions | **Structured** | Numeric records with a clear schema |
| Employee salary data | **Structured** | Tabular, easy to filter and analyse |
| Email messages | **Unstructured** | Free-form text, no predefined format |
| CCTV footage | **Unstructured** | Video — no labels, no order, hard to analyse directly |
| Social media pictures | **Unstructured** | Images — rich content but not directly analysable |

---

**🧠 Mental Map**

```
Data by structure
├── Structured        → organised, labelled, predefined format
│   ├── Easy to filter and analyse
│   ├── Common in spreadsheets and databases
│   └── Examples: sales records, attendance, salary data
│
└── Unstructured      → no labels, no fixed format
    ├── Harder to analyse directly
    ├── Rich in insight when processed
    └── Examples: emails, videos, photos, social media
```

---

**💡 Why this matters**
Most data in the real world is unstructured — emails, support tickets, call recordings, social posts. Before any analysis, you need to know what kind of data you have. Trying to query unstructured data like a spreadsheet is one of the most common early mistakes in analytics.

**✅ RECAP**
- Structured = organised + labelled + predefined format → easy to analyse
- Unstructured = no fixed format → needs processing before analysis
- Neither is better — the type determines the tool, not the value
- Key question before any project: *"What structure does my data have?"*

#### Chapter 1 — Lesson 3: Quantitative vs. qualitative *(Exercise / 100 XP)*

---

**📋 Scenario**
Emma has been organising data at DataPrint and is starting to notice a pattern — some data is all about **numbers and measurements**, while other data describes **qualities and characteristics**.

**Task:** Categorise each item as quantitative or qualitative data.

---

**✅ Solution**

| Item | Classification | Why |
|------|---------------|-----|
| Numerical data | **Quantitative** | By definition — numbers |
| Height of a person | **Quantitative** | Measurable, can be counted and averaged |
| Room temperature | **Quantitative** | Numerical measurement |
| Categorical data | **Qualitative** | By definition — describes categories |
| Language | **Qualitative** | Describes an attribute, cannot be averaged |
| Favourite holiday destination | **Qualitative** | Describes a preference, not a measurement |

---

**🧠 Mental Map**

```
Data by nature
├── Quantitative (numerical)
│   ├── Can be counted, summed, averaged
│   ├── Ideal for calculations and visualisations
│   └── Examples: height, temperature, goals scored, revenue
│
└── Qualitative (categorical)
    ├── Describes attributes or groups
    ├── Useful for spotting patterns and segments
    └── Examples: language, favourite sport, customer feedback
```

---

**💡 Why this matters**
The type of data determines the type of analysis. You cannot calculate an average of "favourite holiday destinations." You cannot segment customers by their height alone. Misidentifying the data type leads to the wrong tool, the wrong chart, and the wrong conclusion.

**✅ RECAP**
- Quantitative = numbers → count, sum, average, visualise
- Qualitative = categories → group, segment, find patterns
- The question to ask first: *"Can I do maths on this?"* If yes → quantitative. If no → qualitative.
- Common mistake: treating qualitative data as numeric (e.g. coding categories as 1, 2, 3 and averaging them)

#### Chapter 1 — Lesson 4: Data context *(Practice / 50 XP)*

---

**📋 Question**
Which of the following is a characteristic of data context?

- The font used to display the data
- ✅ **The origin of the data**
- The color scheme of the data visualisation

**Why the others are wrong:**
- Font and color scheme are presentation choices — they say nothing about the data itself
- The origin of the data tells you *where it came from*, which directly affects how trustworthy and useful it is

**Hint (from DataCamp):** *Consider what provides insight into the background and circumstances of the data. It is not about how the data is displayed or processed, but rather where, when, and why it was collected.*

---

**🧠 Data Context Hierarchy**

```
Data Context  →  central understanding needed for any analysis
├── Source of Data
│   └── Where did the data come from? (origin of the dataset)
├── Conditions of Collection
│   └── How was it gathered? (circumstances during data gathering)
├── Purpose of Collection
│   └── Why was it collected? (intent behind data collection)
└── Relationships Between Data Points
    └── How do the data points connect? (correlations and dependencies)
```

---

**💡 Why this matters**
Two datasets can contain identical numbers but mean completely different things depending on context. A temperature of 37°C recorded in a hospital patient is alarming. The same number recorded as the outdoor temperature in Seville in July is unremarkable. Without context — source, conditions, purpose — data cannot be interpreted correctly.

**✅ RECAP**
- Data context = everything that surrounds the data and makes it interpretable
- The 4 components: source, conditions of collection, purpose, relationships
- Missing context is one of the most dangerous situations in analytics — you can be confident and completely wrong
- Key question: *"Do I know where this data came from, how it was collected, and why?"*

#### Chapter 1 — Lesson 5: The curious case of data growth *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-basics?ex=5) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Understand why data volumes have grown exponentially, how data storage has evolved throughout history, and what drives the need for more data in organisations.

---

**📋 Key concepts covered:**

**1. Exponential data growth**
In 2010: ~2 zettabytes of data created, captured, consumed and stored.
By 2021: 40× more. The growth is not linear — it is exponential, and it is accelerating.

**2. History of data storage**

```
Data storage — evolution over time
├── Nature          → DNA (genetic information preserved since always)
├── Prehistoric     → cave and wall paintings
├── Ancient civilisations → scrolls and papyrus (Chinese, Egyptians, Romans)
│                          used to track financial systems
├── 1890s           → punch cards
├── Mid-20th c.     → magnetic tape, floppy disks
├── Late 20th c.    → CD, hard drive, solid state drive (local storage)
└── Today           → cloud storage — centralised in data centres
                      balances server utilisation, most popular method now
```

**3. Two ice cream shops — why complexity drives data**

| | Shop 1 (gut feeling) | Shop 2 (data-driven) |
|--|---------------------|---------------------|
| Flavours | 3 | 20+ |
| Restocking | Fill whatever ran out | Monitor stock per barcode scan |
| Weather data | No | Yes — predicts sales spikes |
| Pricing | Fixed | Margin per product, adjusted |
| Outcome | Suboptimal — wrong flavours swapped | Optimised revenue long-term |

Key insight: as the complexity of operations increases, so does the amount of data generated and stored.

**4. Real-world examples beyond ice cream**
- **3D manufacturing** — sensors measure beam heat, layer thickness, structural stability
- **Financial institutions** — mortgage applications, fraud detection

---

**💡 Why this matters**
Data growth is not an abstract trend — it is driven by real business decisions. Every time a company chooses to measure something, they generate data. The more complex the operation, the more things worth measuring, the more data. Understanding this connection between business complexity and data volume is foundational for any analyst.

**✅ RECAP**
- Data volume grew 40× between 2010 and 2021 — and it continues to accelerate
- Storage evolved from DNA and cave paintings to cloud — the medium changes, the need to store information does not
- Complexity drives data: more products, more variables, more decisions = more data captured
- Gut feeling leads to suboptimal decisions. Data-driven decisions compound over time.

#### Chapter 1 — Lesson 7: Making the right decisions *(Practice / 50 XP)*

---

**📋 Scenario**
CreaSport makes and sells equipment for racket sports (tennis, squash) — from raw materials to finished products. They are facing two problems: some stores keep running out of stock, and some products are being sold at a loss. CreaSport is just beginning its data journey and wants to make smarter, more competitive decisions using data.

**Task:** Which of the following is the best first step for CreaSport to take as they begin using data to improve their business?

---

**✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| Forecast sales per product per store | ❌ Too advanced | Forecasting requires historical data and a clean baseline — not a first step |
| Investigate supply chain analytics software | ❌ Premature | Buying software before understanding the problem is putting the cart before the horse |
| Close down the least profitable stores | ❌ Destructive | Acting without data on why stores underperform leads to wrong decisions |
| **Calculate profit margin (revenue - cost) per product** | ✅ **Correct** | First, understand where money is being lost. You cannot solve a problem you haven't measured |

**Hint (from DataCamp):** *Becoming data-driven is a digital transformation process that can take multiple years. CreaSport wants to investigate options that solve their two current issues, and work on a long-term plan to optimise their supply chain processes.*

---

**💡 Why this matters**
The first step in any data journey is measurement, not action. Before forecasting, buying software, or closing stores, a business must understand its own numbers. Profit margin per product reveals which products are actually losing money — without that baseline, every other decision is a guess. This is the difference between reacting and reasoning.

**✅ RECAP**
- First step in becoming data-driven: measure what you have, not act on what you feel
- Profit margin = revenue − cost — the simplest possible diagnostic before any strategic decision
- Common mistake: jumping to solutions (software, forecasting, restructuring) before diagnosing the root cause
- Key question: *"Do I understand my current numbers before I try to change them?"*

#### Chapter 1 — Lesson 8: The value of data *(Video / 50 XP)*

**🎯 GOAL — What is this lesson about?**
Understand how data creates value for individuals and organisations across different industries — and why the ability to use data is increasingly a competitive advantage.

---

**📋 Key concepts covered:**

**1. Everyone uses data — individuals and organisations alike**
Individuals use data to manage personal finances, plan their week, track health (e.g. weight, steps). Organisations in every industry — commercial, governmental, non-profit — use data to deal with various challenges.

**2. How organisations use data**
- **Commercial businesses:** ultimate goal is profitability. Data improves decision making by revealing return on investment, optimising processes, and spotting new opportunities
- **Non-profits:** use data for social good or to advance research
- **Customer and employee data:** surveys and interviews generate data that can improve products and employee satisfaction

**3. Data in healthcare**
~1 billion people use wearables (smartwatches, fitness bands, medical devices). With user consent, this data is shared with physicians who use it to detect and predict disorders — enabling **precision medicine**. Patient data also advances pharmaceutical and healthcare research globally.

**4. Data in supply chain**
A supply chain is the sequence of processes involved in production and distribution. Key metrics tracked with data:
- **Average inventory** — number of goods stored in a given period
- **Inventory turnover ratio** — how often inventory is sold and replaced
- **Demand forecasting** — predicting whether the right products will be in stock in time

**5. Data in education**
DataCamp itself uses data to improve courses: analysing user feedback, incorrect answers, and drop-off points helps redesign content. Student performance data identifies where learners struggle and enables personalised learning.

**6. Data as a competitive advantage**
Every organisation's data is unique — competitors cannot replicate the same insights because they do not have the same data. Organisations increasingly compete on how well they use their data to make decisions, not on gut feeling.

---

**💡 Why this matters**
Data is not just a technical asset — it is a strategic one. The same concept (data creates value) appears at every level: a person tracking their weight, a doctor diagnosing a patient, a supply chain manager forecasting demand. The pattern is identical: collect → analyse → decide better. Understanding this pattern across industries is what separates a data analyst from a query runner.

**✅ RECAP**
- Data creates value by improving decision making — at individual, organisational, and industry level
- Four sectors shown: healthcare (precision medicine), supply chain (demand forecasting), education (personalised learning), commercial (profitability + ROI)
- Competitive advantage comes from unique data + the ability to act on it — not from having the most data
- Key question: *"What decision does this data enable — and who benefits from making it better?"*

#### Chapter 1 — Lesson 9: Data-driven vs. gut feelings *(Exercise / 100 XP)*

---

**📋 Scenario**
Emma has been working at DataPrint and is learning that decisions can be made in two very different ways — based on data or based on gut feelings. Her manager is about to launch a new marketing campaign and asks Emma for input. She needs to sort different approaches into data-driven or gut-feeling decisions.

**Task:** Classify each statement as a data-driven or gut-feeling decision.

---

**✅ Solution**

| Statement | Classification | Why |
|-----------|---------------|-----|
| Analysing customer purchase history to decide which products to promote | **Data-driven** | Based on actual behaviour — what customers bought, not what we think they want |
| Using website traffic data to determine the best time to post ads | **Data-driven** | Traffic data provides evidence for timing — not assumption |
| Comparing sales trends before restocking a product | **Data-driven** | Trend analysis = facts about demand over time |
| Choosing a marketing strategy based on personal intuition | **Gut feeling** | No evidence — relies on what feels right |
| Picking a product to promote because it "feels" popular | **Gut feeling** | Perception without measurement — can be completely wrong |
| Deciding on ad timing based on what worked once in the past | **Gut feeling** | One anecdotal instance is not a pattern — no data analysis involved |

**Hint (from DataCamp):** *Intuitive or gut feeling decision-making is how people naturally make decisions without proper tools and procedures, while data-driven choices are purely based on facts.*

---

**💡 Why this matters**
The gut-feeling trap is real — and subtle. "What worked once in the past" feels like experience, but it is a sample of one. "Feels popular" feels like market insight, but it is bias. Data-driven decisions are not better because they are more sophisticated — they are better because they are based on facts rather than the loudest voice or the most recent memory.

**✅ RECAP**
- Data-driven = decision backed by evidence (purchase history, traffic data, sales trends)
- Gut feeling = decision backed by intuition, anecdote, or perception
- The most dangerous gut feelings are the ones that sound like experience
- Key question: *"What is the evidence for this decision — and is it more than one example?"*

#### Chapter 2 — Lesson 1: Data wisdom *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-wisdom?ex=1) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Introduce the DIKW pyramid — the mental model that maps the journey from raw data to actionable wisdom, showing why data alone is never enough to make good decisions.

---

**🧠 Mental Map — The DIKW Pyramid**

```
                    ▲
                   /W\         WISDOM   → make decisions, apply understanding
                  /---\
                 / K   \       KNOWLEDGE → connect pieces, understand relationships
                /-------\
               /    I    \     INFORMATION → data + context = meaning
              /-----------\
             /      D      \   DATA     → raw facts: numbers, words, images, sounds
            /_______________\
```

> ⚠️ **Unconscious bias trap:** The brain automatically tries to add meaning to raw data — before having enough context to do it correctly. This is where bias enters.

---

**🧠 Mental Map — DIKW applied (the James example)**

```
RAW DATA
│  "2"  |  "rain"  |  "15 guests"  |  "James"  |  "park"
│
▼  + context (who, what, where, when)
INFORMATION
│  Age: 2  |  Weather tonight: rain  |  Guest count: 15  |  Venue: park
│
▼  + relationships (how do these connect?)
KNOWLEDGE
│  Rain → muddy park → unsuitable for 15 guests with a 2-year-old
│
▼  + understanding → decision
WISDOM
│  Plan alternative indoor games for tonight's party
```

> *Mental models organise complexity into manageable parts — the DIKW pyramid is one such model. Each level requires refining the one below to extract more meaning.*

---

**💡 Why this matters**
Every data project in business is somewhere on this pyramid. Analysts live in Information. Senior analysts and data scientists work in Knowledge. Business leaders need Wisdom — decisions. The DIKW framework makes explicit what is often left implicit: the distance between having data and being able to act on it. When a stakeholder says *"just show me the data"*, they are asking for the base of the pyramid. When they say *"what should we do?"*, they are asking for the top. Knowing which question is being asked — and which level of the pyramid you are working at — is a core professional skill.

**✅ RECAP**
- DIKW = Data → Information → Knowledge → Wisdom — each level requires refining the one below
- Data alone is never enough to make a decision — you need context, relationships, and understanding
- Unconscious bias lives at the Data level: our brains add meaning before we have enough context
- The distance from data to wisdom is the distance from *having facts* to *knowing what to do*
- Key question: *"Which level of the pyramid is this stakeholder actually asking for?"*

#### Chapter 2 — Lesson 2: Managing mental models *(Practice / 50 XP)*

---

**📋 Scenario**
Mental models help us make sense of complex ideas by organising them into more manageable parts. One popular model is the DIKW framework: Data → Information → Knowledge → Wisdom. At each level, we perform actions to move to the next step, building towards better understanding and decision-making.

**Task:** Which of the following best describes the final step in the DIKW framework?

---

**✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| Collect raw information and apply context to transform it into data | ❌ | This describes moving from Data to Information — the first step, not the last |
| Create wisdom by uncovering connections between pieces of knowledge | ❌ | Uncovering connections is Knowledge — the step before Wisdom |
| **Utilize knowledge to make decisions** | ✅ **Correct** | Wisdom = applying knowledge to act. The final step is not understanding — it is deciding |
| Assemble raw data to serve as the foundation for the entire process | ❌ | This describes the base of the pyramid — Data, not Wisdom |

**Hint (from DataCamp):** *The levels of the DIKW pyramid are: Data → Information → Knowledge → Wisdom.*

---

**💡 Why this matters**
The most common confusion in this exercise is between Knowledge and Wisdom. Knowledge = understanding the relationships between pieces of information. Wisdom = using that understanding to make a decision. You can have complete knowledge and still not act — the moment you act, that is Wisdom. In a business context: knowing that sales drop every January is Knowledge. Deciding to run a January promotion because of that pattern is Wisdom.

**✅ RECAP**
- The final DIKW step is Wisdom = utilizing knowledge to make decisions
- Knowledge and Wisdom are easy to confuse: Knowledge understands relationships, Wisdom acts on them
- Each DIKW level requires an action to move to the next — context, connection, application
- Key question: *"Am I explaining what I know, or deciding what to do?"* — that is the Knowledge/Wisdom boundary

#### Chapter 2 — Lesson 3: DIKW steps *(Exercise / 100 XP)*

---

**📋 Scenario**
The DIKW pyramid focuses on achieving wisdom through various steps throughout the framework.

**Task:** Arrange the steps of DIKW beginning with data at the bottom, unlocking the way to wisdom at the top.

---

**✅ Solution**

| Order | Step | DIKW Level | What gets added |
|-------|------|------------|-----------------|
| 1 (bottom) | Gather measurements and data | **Data** | — starting point |
| 2 | Uncover context for raw data | **Information** | + context |
| 3 | Achieve meaning by connecting different pieces of information | **Knowledge** | + meaning |
| 4 (top) | Apply the findings to guide actions | **Wisdom** | + understanding → decision |

**Hint (from DataCamp):** *Each level requires something to be added to the one that came before. Data becomes information with context. Information becomes knowledge with meaning. Knowledge becomes wisdom with understanding.*

---

**💡 Why this matters**
The three transitions are the three jobs of an analyst. Gathering data → adding context → finding meaning → guiding action. Most analysts spend most of their time in the first two. The organisations that get value from data are the ones that make it all the way to the fourth.

**✅ RECAP**
- Data → Information: add context (when, where, who)
- Information → Knowledge: add meaning (connect the pieces, find relationships)
- Knowledge → Wisdom: add understanding → act on it
- The hardest transition is always the last one — from knowing to deciding

#### Chapter 2 — Lesson 4: Data in decision making *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-wisdom?ex=4) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Introduce the 5-step data-driven decision making process — from asking the right question to making a final decision — and show why this process is more powerful than gut feeling alone.

---

**🧠 Mental Map — The 5-Step Data-Driven Decision Process**

```
QUESTION
│  Identify what you are trying to answer
│  ⚠️ Hardest step — a bad question poisons every step that follows
│
▼
COLLECT
│  Find the right data to answer the question
│  Be deliberate: multiple sources, multiple forms
│  Think ahead — raw or summarised? Decide before collecting
│
▼
PREPARE
│  Clean + transform data into the format analysis requires
│  ⚠️ Can take up to 80% of total project time
│
▼
ANALYSE
│  Transform data into insights using tools
│  (Python, R, Tableau, Power BI, Excel, Google Sheets)
│
▼
DECIDE
│  Interpret results + blend with personal experience and knowledge
│  → data-driven decision = analysis + context + judgment
│
↺  ITERATIVE — decisions feed back into future questions
```

> *Adults face ~35,000 decisions per day. Data cannot help with most of them. But for the ones that matter most, this process turns a tough choice into a clear one.*

---

**💡 Why this matters**
The 5-step process makes explicit what most people skip. They jump straight from "we have a question" to "let's analyse." Without clean data, the analysis is wrong. Without the right question, the analysis is irrelevant. The prepare step alone — routinely underestimated — often takes 80% of the time. Knowing this in advance is the difference between a project that delivers and one that stalls. The iterative nature is also key: every decision is data for the next one.

**✅ RECAP**
- 5 steps: Question → Collect → Prepare → Analyse → Decide
- Step 1 is the hardest — a vague question wastes everything that follows
- Step 3 (Prepare) is the most time-consuming — up to 80% of total time
- The final decision = analysis + personal experience + domain knowledge — not analysis alone
- The process is iterative: decisions generate new data that improves future decisions

#### Chapter 2 — Lesson 5: The right decision *(Exercise / 100 XP)*

---

**📋 Scenario**
Bizvio Corporation can't keep up with high customer service tickets. Tickets are not getting resolved on time and customer satisfaction is dropping. The number of pending service requests is increasing every day. You are brought in to address this situation.

Being data-driven, you break the activity into the five data-driven decision making steps.

**Task:** Order each step of the decision making process to find the right solution.

---

**✅ Solution**

| Order | Step | DDDM Stage |
|-------|------|------------|
| 1 | Create a focused question to uncover information about why customer service tickets are not being properly addressed | **Question** |
| 2 | Decide that the latest customer service data is the best source of data to find insights | **Collect** |
| 3 | Gather the relevant data and highlight any issues or redundancies within the dataset that need to be addressed | **Prepare** |
| 4 | Discover that customers are being contacted multiple times for the same issue, creating multiple tickets that ultimately aren't resolved | **Analyse** |
| 5 | Adjust the process for handling client requests and ticketing, and monitor the outcome to ensure it increases customer satisfaction | **Decide** |

---

**💡 Why this matters**
This is the 5-step process applied to a real business problem. Notice how each step is specific and sequential — you cannot discover the duplicate ticket pattern (step 4) without first cleaning the data (step 3), and you cannot clean the data without first deciding where to source it (step 2). The order is not arbitrary. Skipping steps leads to solving the wrong problem or solving the right problem with bad data.

**✅ RECAP**
- Real-world application of DDDM: Question → Collect → Prepare → Analyse → Decide
- The question must be specific: *why* tickets aren't resolved, not just *that* they aren't
- Prepare step reveals the real insight setup: redundancies in the data point to duplicate tickets
- The decision includes monitoring — data-driven decisions don't end at the action, they close the loop

#### Chapter 2 — Lesson 6: Being data-driven *(Practice / 50 XP)*

---

**📋 Scenario**
Data-driven decision-making gives many advantages to individuals, businesses and organisations — enhancing business growth and operational excellence, optimising customer experience, innovating products and services, and empowering decisions with data.

**Task:** What is one of the main objectives of data-driven decision-making?

---

**✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| To collect as much data as possible | ❌ | More data ≠ better decisions. The goal is the right data for the right question |
| To remove all uncertainty from business decisions | ❌ | Data reduces uncertainty — it cannot eliminate it. Uncertainty is irreducible |
| **To use data insights to guide better and more informed decisions** | ✅ **Correct** | The objective is better decisions, not more data or zero uncertainty |
| To automate every part of the decision-making process | ❌ | Automation is a tool, not the goal. Human judgment remains essential |

**Hint (from DataCamp):** *Think about the advantages of having data to suggest which decisions you should make. Data-driven decision making can help make better decisions.*

---

**💡 Why this matters**
The three wrong answers are three real traps organisations fall into: hoarding data without a question, chasing certainty that doesn't exist, and automating judgment away. The correct answer is deceptively simple — *better and more informed decisions* — but it is the only one that keeps the human in the loop and the question at the centre.

**✅ RECAP**
- The objective of DDDM is better decisions, not more data or zero uncertainty
- Data reduces uncertainty — never eliminates it
- Automation supports decision-making; it does not replace human judgment
- Key question: *"What decision am I trying to improve?"* — not *"How much data can I collect?"*

#### Chapter 2 — Lesson 7: Data as a resource *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-wisdom?ex=7) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Show how raw data must be aggregated to become usable, introduce the data management domains that govern data as an organisational resource, and build the habit of questioning the logic behind metrics and KPIs.

---

**🧠 Mental Map — From Raw Data to Managed Resource**

```
RAW DATA
│  Too large to work with directly
│  Hundreds / millions / billions of records
│
▼  aggregation (sum, average, min, max, mode)
SUMMARISED DATA
│  Distilled into usable form
│  → metrics, benchmarks, KPIs in organisations
│
│  ⚠️ Always ask: what is the business logic behind this metric?
│     What assumptions does this aggregation make?
│
▼  flows through complex pipelines
DATA MANAGEMENT
├── Data Governance   → consistent, trustworthy, not misused
├── Data Quality      → accurate, valid, complete, consistent
├── Data Privacy      → access and use oversight
└── Data Security     → protection of data
```

> *Being curious about aggregations — tracing a report's origin, questioning a KPI's benchmark — unlocks the real value of data.*

---

**💡 Why this matters**
Every number in a dashboard is an aggregation. Every KPI is a business decision encoded as a formula. Analysts who treat metrics as given facts miss half the job. The metric is only as trustworthy as the logic behind it — and that logic is often undocumented, assumed, or wrong. The four data management domains (governance, quality, privacy, security) exist because organisations learned this the hard way. Knowing they exist — and what each one protects — is the baseline for any serious data role.

**✅ RECAP**
- Raw data is too large to use directly — aggregation is the first transformation step
- Common aggregations: sum, average, min, max, mode → appear as metrics, benchmarks, KPIs
- Critical habit: always ask *what is the business logic behind this aggregation?*
- Data management = the field that governs data pipelines across an organisation
- Four domains: Governance (trust), Quality (accuracy), Privacy (access), Security (protection)
- Every analyst is responsible for data being treated as a valuable resource — not just data teams

#### Chapter 2 — Lesson 8: Aggregating information *(Practice / 50 XP)*

---

**📋 Scenario**
Aggregations help individuals and organisations zoom in on specific parts of a dataset, making it easier to discover insights. However, it's important to approach them thoughtfully, especially the first time you use one.

**Task:** Which of the following is most important to consider when working with data aggregation?

---

**✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| Aggregations tell you everything you need to know about a dataset | ❌ | Aggregations summarise — they also hide. An average conceals the distribution beneath it |
| Aggregations work the same way across all datasets | ❌ | Organisations have varying rules that cause small but important differences in how metrics are calculated |
| **It is important to understand the business logic behind an aggregation** | ✅ **Correct** | The same metric name can mean different things in different organisations — the logic behind it is what gives it meaning |
| Aggregations are different from KPIs and metrics | ❌ | KPIs and metrics are aggregations — they are the same thing expressed in business language |

**Hint (from DataCamp):** *Organisations can have varying rules that may result in small but important changes to common metrics — understanding these rules is vital.*

---

**💡 Why this matters**
"Revenue" sounds universal. But is it gross or net? Does it include returns? Is it recognised on invoice date or payment date? Two analysts at the same company can produce different numbers using the same word because the business logic differs. This is not a data quality issue — it is a business logic issue. And it can only be caught by someone who asks the question.

**✅ RECAP**
- Aggregations summarise data — but they also hide what's beneath the summary
- The same aggregation (e.g. "revenue") can be calculated differently across teams or organisations
- Most important question when encountering a metric: *"How exactly is this calculated?"*
- KPIs and metrics ARE aggregations — just expressed in business language

#### Chapter 2 — Lesson 9: Data management *(Exercise / 100 XP)*

---

**📋 Scenario**
Data is a highly valuable asset to every organisation. Managing that flow of data requires dedicated efforts from many different individuals to unlock the potential within data streams.

**Task:** Match each statement to the correct data management domain: Data Governance, Data Quality, Data Security and Privacy, Data Management.

---

**✅ Solution**

| Statement | Domain | Core responsibility |
|-----------|--------|-------------------|
| Ensures that data is consistent, trustworthy and is used correctly across an organisation | **Data Governance** | How data is used — policy and trust |
| Responsible for validating that data is accurate, complete and consistent | **Data Quality** | The integrity of the data itself |
| Regulates proper access, use and protection of data | **Data Security and Privacy** | Who can access data and how it is protected |
| Oversees and coordinates all the subdomains into one unifying structure | **Data Management** | The umbrella that holds all domains together |

**Hint (from DataCamp):** *The sub-domains of data management are each responsible for a different area of managing data. Data Governance focuses on how data is used. Data Quality aims to deliver accurate data free from errors. Data Security and Privacy manage who has access and how data is protected.*

---

**💡 Why this matters**
These four domains appear in job descriptions, data strategy documents, and compliance frameworks. An analyst who can name them and explain what each one protects will stand out in any interview. More practically: when something goes wrong with data in an organisation, one of these four domains has failed. Knowing which one tells you where to look.

**✅ RECAP**
- Data Governance = consistent, trustworthy, correctly used — the policy layer
- Data Quality = accurate, valid, complete, consistent — the integrity layer
- Data Security and Privacy = access, use, protection — the safety layer
- Data Management = the umbrella that coordinates all three subdomains
- When data goes wrong, ask: *which domain failed?*
