#### Chapter 1 — Lesson 1: Data basics *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-basics?ex=1) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Establish what data is, where it comes from, and how to classify it — the conceptual foundation before any analysis begins.

---

**🧠 Mental Map — Data Classification: Two Axes**

```
DATA
├── by STRUCTURE
│   ├── Structured   → organised, labelled, predefined format
│   │   └── Examples: spreadsheets, databases, employee records
│   └── Unstructured → no labels, no fixed order
│       └── Examples: videos, photos, emails, social media posts
│
└── by NATURE
    ├── Quantitative → numerical — can be counted, summed, averaged
    │   └── Examples: goals scored, height, temperature, revenue
    └── Qualitative  → categorical — describes attributes or groups
        └── Examples: favourite sport, language, customer feedback
```

> *"Data" comes from the Latin datum — meaning "fact." Raw facts without context are noise. Context around data — when, where, how it was collected — is called **metadata**, and it is what makes data interpretable.*

---

**💡 Why this matters**
An analyst who cannot classify data quickly reaches for the wrong tool. Applying a statistical average to qualitative categories, or trying to query unstructured data like a table — both are real mistakes that start here, at the classification step. These two axes (structure and nature) are the first filter before any project begins. Every other decision — which database, which chart, which analysis method — follows from this.

**✅ RECAP**
- Data = facts. Facts need context to have meaning.
- Two classification axes: structure (structured/unstructured) and nature (quantitative/qualitative)
- Metadata = the context around data (when, where, how it was collected)
- Neither axis is better — the right type depends on the question being asked
- The most common mistake: treating data as self-explanatory. It never is.

#### Chapter 1 — Lesson 2: Structured vs. unstructured *(Exercise / 100 XP)*

---

**📋 Scenario**
Emma just started her new job at DataPrint, an online store for printer gear. In her first week, she discovered loads of data and needs to sort it into **structured** or **unstructured** categories.

**Task:** Classify each item as structured or unstructured data.

---

**✅ Solution**

| Item | Classification | Why |
|------|---------------|-----|
| Customer information (Name, Age, ID) | **Structured** | Organised in rows and columns, predefined format |
| Financial transactions | **Structured** | Numeric records with a clear schema |
| Employee salary data | **Structured** | Tabular, easy to filter and analyse |
| Email messages | **Unstructured** | Free-form text, no predefined format |
| CCTV footage | **Unstructured** | Video — no labels, no order, hard to analyse directly |
| Social media pictures | **Unstructured** | Images — rich content but not directly analysable |

---

**🧠 Mental Map**

```
Data by structure
├── Structured        → organised, labelled, predefined format
│   ├── Easy to filter and analyse
│   ├── Common in spreadsheets and databases
│   └── Examples: sales records, attendance, salary data
│
└── Unstructured      → no labels, no fixed format
    ├── Harder to analyse directly
    ├── Rich in insight when processed
    └── Examples: emails, videos, photos, social media
```

---

**💡 Why this matters**
Most data in the real world is unstructured — emails, support tickets, call recordings, social posts. Before any analysis, you need to know what kind of data you have. Trying to query unstructured data like a spreadsheet is one of the most common early mistakes in analytics.

**✅ RECAP**
- Structured = organised + labelled + predefined format → easy to analyse
- Unstructured = no fixed format → needs processing before analysis
- Neither is better — the type determines the tool, not the value
- Key question before any project: *"What structure does my data have?"*

#### Chapter 1 — Lesson 3: Quantitative vs. qualitative *(Exercise / 100 XP)*

---

**📋 Scenario**
Emma has been organising data at DataPrint and is starting to notice a pattern — some data is all about **numbers and measurements**, while other data describes **qualities and characteristics**.

**Task:** Categorise each item as quantitative or qualitative data.

---

**✅ Solution**

| Item | Classification | Why |
|------|---------------|-----|
| Numerical data | **Quantitative** | By definition — numbers |
| Height of a person | **Quantitative** | Measurable, can be counted and averaged |
| Room temperature | **Quantitative** | Numerical measurement |
| Categorical data | **Qualitative** | By definition — describes categories |
| Language | **Qualitative** | Describes an attribute, cannot be averaged |
| Favourite holiday destination | **Qualitative** | Describes a preference, not a measurement |

---

**🧠 Mental Map**

```
Data by nature
├── Quantitative (numerical)
│   ├── Can be counted, summed, averaged
│   ├── Ideal for calculations and visualisations
│   └── Examples: height, temperature, goals scored, revenue
│
└── Qualitative (categorical)
    ├── Describes attributes or groups
    ├── Useful for spotting patterns and segments
    └── Examples: language, favourite sport, customer feedback
```

---

**💡 Why this matters**
The type of data determines the type of analysis. You cannot calculate an average of "favourite holiday destinations." You cannot segment customers by their height alone. Misidentifying the data type leads to the wrong tool, the wrong chart, and the wrong conclusion.

**✅ RECAP**
- Quantitative = numbers → count, sum, average, visualise
- Qualitative = categories → group, segment, find patterns
- The question to ask first: *"Can I do maths on this?"* If yes → quantitative. If no → qualitative.
- Common mistake: treating qualitative data as numeric (e.g. coding categories as 1, 2, 3 and averaging them)

#### Chapter 1 — Lesson 4: Data context *(Practice / 50 XP)*

---

**📋 Question**
Which of the following is a characteristic of data context?

- The font used to display the data
- ✅ **The origin of the data**
- The color scheme of the data visualisation

**Why the others are wrong:**
- Font and color scheme are presentation choices — they say nothing about the data itself
- The origin of the data tells you *where it came from*, which directly affects how trustworthy and useful it is

**Hint (from DataCamp):** *Consider what provides insight into the background and circumstances of the data. It is not about how the data is displayed or processed, but rather where, when, and why it was collected.*

---

**🧠 Data Context Hierarchy**

```
Data Context  →  central understanding needed for any analysis
├── Source of Data
│   └── Where did the data come from? (origin of the dataset)
├── Conditions of Collection
│   └── How was it gathered? (circumstances during data gathering)
├── Purpose of Collection
│   └── Why was it collected? (intent behind data collection)
└── Relationships Between Data Points
    └── How do the data points connect? (correlations and dependencies)
```

---

**💡 Why this matters**
Two datasets can contain identical numbers but mean completely different things depending on context. A temperature of 37°C recorded in a hospital patient is alarming. The same number recorded as the outdoor temperature in Seville in July is unremarkable. Without context — source, conditions, purpose — data cannot be interpreted correctly.

**✅ RECAP**
- Data context = everything that surrounds the data and makes it interpretable
- The 4 components: source, conditions of collection, purpose, relationships
- Missing context is one of the most dangerous situations in analytics — you can be confident and completely wrong
- Key question: *"Do I know where this data came from, how it was collected, and why?"*

#### Chapter 1 — Lesson 5: The curious case of data growth *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-basics?ex=5) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Understand why data volumes have grown exponentially, how data storage has evolved throughout history, and what drives the need for more data in organisations.

---

**🧠 Mental Map — Data Storage: From Cave Walls to Cloud**

```
DATA STORAGE — EVOLUTION
├── Nature          → DNA (genetic information — preserved since always)
├── Prehistoric     → cave and wall paintings
├── Ancient civilisations → scrolls and papyrus
│                          used to track financial systems
├── 1890s           → punch cards
├── Mid-20th c.     → magnetic tape, floppy disks
├── Late 20th c.    → CD, hard drive, solid state drive (local storage)
└── Today           → cloud storage — centralised in data centres
                       balances server utilisation; most popular method now
```

> *In 2010: ~2 zettabytes of data existed globally. By 2021: 40× more. The growth is exponential — and accelerating. The medium changes, the need to store information does not.*

**Why complexity drives data — two ice cream shops:**

| | Shop 1 (gut feeling) | Shop 2 (data-driven) |
|--|---------------------|---------------------|
| Flavours | 3 | 20+ |
| Restocking | Fill whatever ran out | Monitor stock per barcode scan |
| Weather data | No | Yes — predicts sales spikes |
| Pricing | Fixed | Margin per product, adjusted |
| Outcome | Suboptimal | Optimised revenue long-term |

The more complex the operation, the more things worth measuring, the more data generated and stored. This pattern holds in 3D manufacturing (sensors per beam), financial institutions (fraud detection), healthcare (wearables), and beyond.

---

**💡 Why this matters**
Data growth is not an abstract trend — it is driven by real business decisions. Every time a company chooses to measure something, they generate data. Understanding the link between business complexity and data volume prepares an analyst for the reality of modern data work: the datasets are large, the storage choices matter, and the question "should we be storing this?" is as important as "can we?"

**✅ RECAP**
- Data volume grew 40× between 2010 and 2021 — exponential, not linear
- Storage evolved from DNA and cave paintings to cloud — medium changes, need to store does not
- Complexity drives data: more products, more variables, more decisions = more data captured
- Gut feeling leads to suboptimal decisions. Data-driven decisions compound over time.

#### Chapter 1 — Lesson 7: Making the right decisions *(Practice / 50 XP)*

---

**📋 Scenario**
CreaSport makes and sells equipment for racket sports (tennis, squash) — from raw materials to finished products. They are facing two problems: some stores keep running out of stock, and some products are being sold at a loss. CreaSport is just beginning its data journey and wants to make smarter, more competitive decisions using data.

**Task:** Which of the following is the best first step for CreaSport to take as they begin using data to improve their business?

---

**✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| Forecast sales per product per store | ❌ Too advanced | Forecasting requires historical data and a clean baseline — not a first step |
| Investigate supply chain analytics software | ❌ Premature | Buying software before understanding the problem is putting the cart before the horse |
| Close down the least profitable stores | ❌ Destructive | Acting without data on why stores underperform leads to wrong decisions |
| **Calculate profit margin (revenue - cost) per product** | ✅ **Correct** | First, understand where money is being lost. You cannot solve a problem you haven't measured |

**Hint (from DataCamp):** *Becoming data-driven is a digital transformation process that can take multiple years. CreaSport wants to investigate options that solve their two current issues, and work on a long-term plan to optimise their supply chain processes.*

---

**💡 Why this matters**
The first step in any data journey is measurement, not action. Before forecasting, buying software, or closing stores, a business must understand its own numbers. Profit margin per product reveals which products are actually losing money — without that baseline, every other decision is a guess. This is the difference between reacting and reasoning.

**✅ RECAP**
- First step in becoming data-driven: measure what you have, not act on what you feel
- Profit margin = revenue − cost — the simplest possible diagnostic before any strategic decision
- Common mistake: jumping to solutions (software, forecasting, restructuring) before diagnosing the root cause
- Key question: *"Do I understand my current numbers before I try to change them?"*

#### Chapter 1 — Lesson 8: The value of data *(Video / 50 XP)*

---

**🎯 GOAL — What is this lesson about?**
Understand how data creates value for individuals and organisations across different industries — and why the ability to use data is increasingly a competitive advantage.

---

Everyone uses data — individuals track finances, health, and weekly plans. Organisations across commercial, governmental, and non-profit sectors use data to deal with different challenges. The pattern at every level is identical: collect → analyse → decide better.

**In healthcare**, ~1 billion people use wearables. With consent, this data is shared with physicians who use it to detect and predict disorders — enabling **precision medicine**. Patient data also advances pharmaceutical and healthcare research globally.

**In supply chain**, key metrics tracked with data: average inventory (goods stored per period), inventory turnover ratio (how often inventory is replaced), and demand forecasting (predicting whether the right products will be in stock in time).

**In education**, DataCamp itself analyses user feedback, incorrect answers, and drop-off points to redesign content. Student performance data identifies where learners struggle and enables personalised learning.

**In commercial business**, the ultimate goal is profitability. Data improves decision making by revealing return on investment, optimising processes, and spotting new opportunities. Every organisation's data is unique — competitors cannot replicate the same insights because they do not have the same data.

---

**💡 Why this matters**
Data is not just a technical asset — it is a strategic one. The same concept (data creates value) appears at every level: a person tracking their weight, a doctor diagnosing a patient, a supply chain manager forecasting demand. Understanding this pattern across industries is what separates a data analyst from a query runner. Competitive advantage no longer comes from having the most data — it comes from using it better.

**✅ RECAP**
- Data creates value by improving decision making — at individual, organisational, and industry level
- Four sectors: healthcare (precision medicine), supply chain (demand forecasting), education (personalised learning), commercial (profitability + ROI)
- Competitive advantage = unique data + the ability to act on it
- Key question: *"What decision does this data enable — and who benefits from making it better?"*

#### Chapter 1 — Lesson 9: Data-driven vs. gut feelings *(Exercise / 100 XP)*

---

**📋 Scenario**
Emma has been working at DataPrint and is learning that decisions can be made in two very different ways — based on data or based on gut feelings. Her manager is about to launch a new marketing campaign and asks Emma for input. She needs to sort different approaches into data-driven or gut-feeling decisions.

**Task:** Classify each statement as a data-driven or gut-feeling decision.

---

**✅ Solution**

| Statement | Classification | Why |
|-----------|---------------|-----|
| Analysing customer purchase history to decide which products to promote | **Data-driven** | Based on actual behaviour — what customers bought, not what we think they want |
| Using website traffic data to determine the best time to post ads | **Data-driven** | Traffic data provides evidence for timing — not assumption |
| Comparing sales trends before restocking a product | **Data-driven** | Trend analysis = facts about demand over time |
| Choosing a marketing strategy based on personal intuition | **Gut feeling** | No evidence — relies on what feels right |
| Picking a product to promote because it "feels" popular | **Gut feeling** | Perception without measurement — can be completely wrong |
| Deciding on ad timing based on what worked once in the past | **Gut feeling** | One anecdotal instance is not a pattern — no data analysis involved |

**Hint (from DataCamp):** *Intuitive or gut feeling decision-making is how people naturally make decisions without proper tools and procedures, while data-driven choices are purely based on facts.*

---

**💡 Why this matters**
The gut-feeling trap is real — and subtle. "What worked once in the past" feels like experience, but it is a sample of one. "Feels popular" feels like market insight, but it is bias. Data-driven decisions are not better because they are more sophisticated — they are better because they are based on facts rather than the loudest voice or the most recent memory.

**✅ RECAP**
- Data-driven = decision backed by evidence (purchase history, traffic data, sales trends)
- Gut feeling = decision backed by intuition, anecdote, or perception
- The most dangerous gut feelings are the ones that sound like experience
- Key question: *"What is the evidence for this decision — and is it more than one example?"*

#### Chapter 2 — Lesson 1: Data wisdom *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-wisdom?ex=1) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Introduce the DIKW pyramid — the mental model that maps the journey from raw data to actionable wisdom, showing why data alone is never enough to make good decisions.

---

**🧠 Mental Map — The DIKW Pyramid**

```
                    ▲
                   /W\         WISDOM   → make decisions, apply understanding
                  /---\
                 / K   \       KNOWLEDGE → connect pieces, understand relationships
                /-------\
               /    I    \     INFORMATION → data + context = meaning
              /-----------\
             /      D      \   DATA     → raw facts: numbers, words, images, sounds
            /_______________\
```

> ⚠️ **Unconscious bias trap:** The brain automatically tries to add meaning to raw data — before having enough context to do it correctly. This is where bias enters.

---

**🧠 Mental Map — DIKW applied (the James example)**

```
RAW DATA
│  "2"  |  "rain"  |  "15 guests"  |  "James"  |  "park"
│
▼  + context (who, what, where, when)
INFORMATION
│  Age: 2  |  Weather tonight: rain  |  Guest count: 15  |  Venue: park
│
▼  + relationships (how do these connect?)
KNOWLEDGE
│  Rain → muddy park → unsuitable for 15 guests with a 2-year-old
│
▼  + understanding → decision
WISDOM
│  Plan alternative indoor games for tonight's party
```

> *Mental models organise complexity into manageable parts — the DIKW pyramid is one such model. Each level requires refining the one below to extract more meaning.*

---

**💡 Why this matters**
Every data project in business is somewhere on this pyramid. Analysts live in Information. Senior analysts and data scientists work in Knowledge. Business leaders need Wisdom — decisions. The DIKW framework makes explicit what is often left implicit: the distance between having data and being able to act on it. When a stakeholder says *"just show me the data"*, they are asking for the base of the pyramid. When they say *"what should we do?"*, they are asking for the top. Knowing which question is being asked — and which level of the pyramid you are working at — is a core professional skill.

**✅ RECAP**
- DIKW = Data → Information → Knowledge → Wisdom — each level requires refining the one below
- Data alone is never enough to make a decision — you need context, relationships, and understanding
- Unconscious bias lives at the Data level: our brains add meaning before we have enough context
- The distance from data to wisdom is the distance from *having facts* to *knowing what to do*
- Key question: *"Which level of the pyramid is this stakeholder actually asking for?"*

#### Chapter 2 — Lesson 2: Managing mental models *(Practice / 50 XP)*

---

**📋 Scenario**
Mental models help us make sense of complex ideas by organising them into more manageable parts. One popular model is the DIKW framework: Data → Information → Knowledge → Wisdom. At each level, we perform actions to move to the next step, building towards better understanding and decision-making.

**Task:** Which of the following best describes the final step in the DIKW framework?

---

**✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| Collect raw information and apply context to transform it into data | ❌ | This describes moving from Data to Information — the first step, not the last |
| Create wisdom by uncovering connections between pieces of knowledge | ❌ | Uncovering connections is Knowledge — the step before Wisdom |
| **Utilize knowledge to make decisions** | ✅ **Correct** | Wisdom = applying knowledge to act. The final step is not understanding — it is deciding |
| Assemble raw data to serve as the foundation for the entire process | ❌ | This describes the base of the pyramid — Data, not Wisdom |

**Hint (from DataCamp):** *The levels of the DIKW pyramid are: Data → Information → Knowledge → Wisdom.*

---

**💡 Why this matters**
The most common confusion in this exercise is between Knowledge and Wisdom. Knowledge = understanding the relationships between pieces of information. Wisdom = using that understanding to make a decision. You can have complete knowledge and still not act — the moment you act, that is Wisdom. In a business context: knowing that sales drop every January is Knowledge. Deciding to run a January promotion because of that pattern is Wisdom.

**✅ RECAP**
- The final DIKW step is Wisdom = utilizing knowledge to make decisions
- Knowledge and Wisdom are easy to confuse: Knowledge understands relationships, Wisdom acts on them
- Each DIKW level requires an action to move to the next — context, connection, application
- Key question: *"Am I explaining what I know, or deciding what to do?"* — that is the Knowledge/Wisdom boundary

#### Chapter 2 — Lesson 3: DIKW steps *(Exercise / 100 XP)*

---

**📋 Scenario**
The DIKW pyramid focuses on achieving wisdom through various steps throughout the framework.

**Task:** Arrange the steps of DIKW beginning with data at the bottom, unlocking the way to wisdom at the top.

- Step 3 (Prepare) is the most time-consuming — up to 80% of total time

#### Chapter 3 — Lesson 1: Data ethics and privacy *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-in-depth?ex=1) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Define data ethics as the framework that governs how organisations collect, store, and use data responsibly — and introduce the five key principles that make it operational.

---

**🧠 Mental Map — The 5 Principles of Data Ethics**

```
DATA ETHICS
│  How organisations ethically collect, store, and use data
│  Goal: protect customers + protect the organisation from legal risk
│
├── PERMISSION
│   Collect only with explicit user consent
│   Users must be in control — data given, not taken
│
├── TRANSPARENCY
│   Tell users how data will be collected, stored, and used
│   ⚠️ Lack of transparency → reputation damage + legal exposure
│
├── PRIVACY  ← specific to PII
│   PII = name, birthdate, ID number, phone number
│   │
│   └── Two ways to protect it:
│       ├── Limit what you collect (most effective)
│       │   ⚠️ Phishing, ransomware, social media oversharing
│       └── Pseudo-anonymisation
│           Remove PII from dataset → patterns remain, individuals disappear
│
├── INTENTION
│   Before collecting: "Should I be doing this with data?"
│   "How will this benefit customers or society?"
│   If no good answer → reconsider
│
└── OUTCOME
    Monitor actual results — good intentions still cause harm
    Protect vulnerable populations
    ⚠️ Intent ≠ impact
```

> *Data ethics is constantly evolving — there is no single consensus. The 5-principle framework (Harvard Business School, 2021) is one structured approach, not the final word.*

---

**💡 Why this matters**
In any data role, ethics is not abstract — it shows up in every collection project, every dataset you build, every dashboard you publish. GDPR, CCPA, and similar regulations are legal codifications of these five principles. An analyst who cannot explain what PII is, or articulate the difference between permission and transparency, is a compliance risk. One who can is a safer, more trusted hire. The most effective protection is also the simplest: collect less.

**✅ RECAP**
- 5 principles: Permission, Transparency, Privacy, Intention, Outcome
- Privacy is specifically about PII — personally identifiable information
- Most effective privacy protection: limit what you collect in the first place
- Pseudo-anonymisation = remove PII, keep the patterns — individuals disappear, insights remain
- Key question before any data project: *"Should I be doing this? How does it benefit people?"*
- Intent ≠ impact — good intentions must be monitored against actual outcomes

#### Chapter 3 — Lesson 5: Data life cycle *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-in-depth?ex=5) *(requires login)*

**🎯 GOAL — What is this lesson about?**
Introduce the data life cycle as a governance framework — six sequential stages that take data from its first question through to destruction or archival, and explain why each stage exists.

---

**🧠 Mental Map — The Data Life Cycle**

```
PLAN & COLLECT
│  Define the business question first — it drives all other decisions
│  Decide: what data, who owns it, how it will be managed
│  Sources: surveys, experiments, sensor readings, external datasets
│
▼
STORE & MANAGE
│  Data lands in a database or data warehouse
│  ⚠️ Address PII and sensitive data handling here — not later
│
▼
CLEAN & PROCESS
│  Format, fix missing values, remove errors, transform into usable form
│  ⚠️ Largest effort in the entire life cycle — often underestimated
│
▼
ANALYSE & VISUALISE
│  Extract insights from cleaned data
│  Methods: statistical analysis, machine learning, BI tools
│  Visualisation makes insights interpretable
│
▼
SHARE
│  Communicate findings to stakeholders
│  Forms: dashboards, reports, papers, presentations, published datasets
│  ⚠️ Analysis that is never shared has zero value
│
▼
ARCHIVE OR DESTROY
│  Does the data need to be kept for future use?
│  ├── Archive → backup, documentation, digital preservation
│  └── Destroy → permanent deletion — protects privacy, frees resources
│
↺  The end feeds back into new questions — the cycle is iterative
```

> *The planning stage is the most consequential — a vague business question corrupts every stage that follows. "What do we want to know?" must be answered before any data is touched.*

---

**💡 Why this matters**
The data life cycle is the backbone of every data project, whether you are building a dashboard or training a machine learning model. Every stage has a cost — time, storage, compute, human attention. Skipping or rushing a stage does not eliminate the cost; it pushes it downstream, where it multiplies. Cleaning a poorly stored dataset takes far longer than storing it correctly in the first place. Sharing results that were never planned for an audience is harder than designing for communication from the start. The life cycle is a map for avoiding those compounding costs.

**✅ RECAP**
- 6 stages: Plan & Collect → Store & Manage → Clean & Process → Analyse & Visualise → Share → Archive or Destroy
- Stage 1 (Plan) is the most consequential — the business question shapes everything that follows
- Stage 3 (Clean & Process) is the most time-consuming — the largest effort in the cycle
- Stage 5 (Share) is the most overlooked — unshared analysis has no value
- Stage 6 has two outcomes: Archive (keep with documentation) or Destroy (permanent deletion for privacy)
- The cycle is iterative — every answer generates new questions

| 6 | **Archive or Destroy** | Since the collected data doesn't contain any PII or other personal data, Rob decides to make the datasets publicly available |

**Hint (from DataCamp):** *The data life cycle starts with identifying and collecting data, and ends with its long-term storage or removal.*

---

**💡 Why this matters**
This exercise is the life cycle made tangible. A few things worth noting: Stage 2 (CSV → SQL database) is a storage and management decision — not a cleaning step, even though it involves transformation. Stage 5 (presenting at a conference to mixed audiences) highlights that sharing requires communication skill, not just analysis. Stage 6 ends with making the data public rather than deleting it — because no PII means archiving is safe and adds value to other researchers. The end-of-life decision is always driven by what the data contains.

**✅ RECAP**
- The 6 phases applied: Plan → Store → Clean → Analyse → Share → Archive/Destroy
- Converting CSVs to SQL = storage decision, not a cleaning step
- Sharing to mixed audiences = communication design challenge, not just export to PDF
- No PII → archive and publish; contains PII → destroy or anonymise before archiving
- The end-of-life decision is determined by what the data contains, not by convenience

#### Chapter 3 — Lesson 9: Common data mistakes *(Video / 50 XP)*

> 🎬 [Watch on DataCamp](https://campus.datacamp.com/courses/introduction-to-data/data-in-depth?ex=9) *(requires login)*

---

**🎯 GOAL — What is this lesson about?**
Name the four most common data mistakes, map each to the stage of the life cycle where it originates, and show how each one cascades into incorrect conclusions downstream.

---

**🧠 Mental Map — 4 Common Data Mistakes**

```
PLAN & COLLECT
│
├── ❌ MISTAKE 1: Vague problem definition
│   "Did you buy anything last month?" — too broad to yield useful data
│   → Risk: wrong data collected, wrong question answered, wrong conclusions
│   ✓ Fix: specific, targeted questions — "Which payment method did you use?"
│
├── ❌ MISTAKE 2: Insufficient or biased data
│   Target: elderly people. Actual respondents: mostly young adults.
│   → This is data bias — responses not representative of target audience
│   ✓ Fix: design collection to reach the right population
│   ⚠️ Even correct data still needs cleaning before analysis
│
ANALYSE & VISUALISE
│
├── ❌ MISTAKE 3: Lack of appropriate analysis / missing context
│   Steep decline in contactless payments → "elderly people avoid it"
│   Reality: payment terminals had technical issues that week
│   → Correlation mistaken for causation; context ignored
│   Other forms: wrong aggregations, incorrect calculations
│
SHARE
│
└── ❌ MISTAKE 4: Poor communication of results
    Forms this takes:
    ├── Technical analysis presented to non-technical audience — no business relevance shown
    ├── Cherry-picking data points to support a predetermined conclusion
    ├── Misleading chart types
    └── Missing labels, legends, axis titles, colours → misinterpretation
    ⚠️ Poor communication makes all previous work worthless — or actively harmful
```

> *Planning ahead reduces mistakes — but does not eliminate them. Each stage of the life cycle has its own failure mode.*

---

**💡 Why this matters**
These four mistakes are not edge cases — they are the default. Most real-world data projects fail at one of these four points. Vague questions are written because no one pushed back on the brief. Biased data goes unnoticed because no one checked who actually responded. Missing context produces confident wrong conclusions. Poor communication means the right answer never reaches the person who needs it. The life cycle is the diagnostic framework — when something goes wrong, one of these four stages is where the failure originated.

**✅ RECAP**
- 4 mistakes, 4 stages: vague question (Plan) → biased data (Collect) → missing context (Analyse) → poor communication (Share)
- Data bias: when respondents are not representative of the target group
- Correlation ≠ causation — always ask: *what else could explain this pattern?*
- Cherry-picking and misleading charts are as dangerous as incorrect analysis
- The most valuable part of the life cycle is communication — the most commonly skipped

#### Chapter 3 — Lesson 10: Lack of explanation *(Practice / 50 XP)*

---

**📋 Scenario**
A table shows exam results for 5 students (John, David, William, Elizabeth, Sarah) across Math, Physics, Chemistry, and Biology. A pie chart is shown representing each student's total score as a share of the combined total.

The lack of clear communication is one of the major hurdles in interpreting data. If you don't know exactly what the data is about, you can't draw any conclusions.

**Task:** Identify the correct statement about the pie chart shown.

---

**✅ Solution**

| Option | Verdict | Why |
|--------|---------|-----|
| Different colours will make the pie chart more understandable | ❌ | The chart already uses different colours per student — more colour does not fix the deeper problem of chart type mismatch |
| **The data would be better represented with a line or bar chart** | ✅ **Correct** | A pie chart shows part-to-whole relationships. Exam totals compared across students is a ranking/comparison — a bar chart is the right tool |
| Adding more labels will show how the pie chart segments align with the data shown in the table | ❌ | Labels patch a symptom, not the cause — the wrong chart type is the problem, not missing labels |

**Hint (from DataCamp):** *Data visualisations should be as self-explanatory as possible, and must contain all the relevant details so that they are understandable to the audience.*

---

**💡 Why this matters**
Chart type selection is a communication decision, not a cosmetic one. A pie chart answers the question: *"What share of the whole does each part represent?"* Five students' exam totals don't form a meaningful whole — no one is asking "what percentage of the combined class total did John score?" The real question is "who scored highest?" — and that is a bar chart. Using the wrong chart type does not just look bad; it implies a question the data was never asked to answer.

**✅ RECAP**
- Pie charts = part-to-whole relationships. Use them only when the "whole" is meaningful
- Bar/line charts = comparisons and rankings across categories or time
- Adding labels to the wrong chart type fixes presentation, not the underlying mismatch
- Key question before any visualisation: *"What question is this chart answering?"*
- A chart that cannot be understood without explanation has already failed

#### Chapter 3 — Lesson 11: Spot the mistakes *(Exercise / 50 XP)*

**📋 Scenario**

| Statement | Verdict | Mistake type |

| **An analysis that calculates the sum of all ages per age category** | ❌ **Mistake** | Wrong aggregation — summing ages within age categories is meaningless. Count or average would be appropriate |

**Hint (from DataCamp):** *Mistakes often involve unclear visuals, incorrect or incomplete data, or drawing conclusions without proper analysis.*

---

**💡 Why this matters**
Two of the four statements describe problems that were *caught and fixed* — and those are not mistakes. This distinction matters: the goal is not to never encounter problems, it is to catch them before they corrupt the output. The two actual mistakes are subtle: summing ages sounds like analysis, but produces numbers with no meaning. Presenting conclusions on half the agreed sample sounds like progress, but the agreement on sample size existed for a reason — the conclusions are not valid without it.

**✅ RECAP**
- Wrong aggregation: sum of ages per age group = meaningless number — count or average is appropriate

- Catching a missing legend before presenting = correct process, not a mistake
- Almost missing context but including it = correct process, not a mistake
- The standard is not "no problems encountered" — it is "problems caught before they cause harm"
